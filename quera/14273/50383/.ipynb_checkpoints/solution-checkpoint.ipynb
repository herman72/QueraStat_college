            مرور و مقدمه
        </font>
		<p></p>
        <hr>
        تا به این جا با مفهوم متغیر تصادفی آشنا شدیم و مثال هایی از آن را چه در حالت گسسته و چه در حالت پیوسته بررسی کردیم . علاوه بر آن توزیع های مختلفی را برای این متغیرهای تصادفی مرور کردیم . در این فصل با حالت هایی روبرو خواهیم شد که نیازمند داشتن یک توزیع مناسب از چند متغیر تصادفی به شکل توام یا همزمان هستیم . مثال های متعددی در اطرافمان از این دست هستند که در این جا برای درک بهتر به برخی از آنها می پردازیم .
        <ul>
            <li>
                احتمال سرطان داشتن را می‌خواهیم بررسی کنیم با توجه به سیگاری بودن و وزن و خوراکی های مورد مصرف شخص
            </li>
            <li>
                میخواهیم میزان فروش یک فیلم سینمایی را بررسی کنیم اما با توجه به بازیگران فیلم و سطح اگاهی و علاقه مندی افراد یک جامعه
            </li>
        </ul>
        <br>
           سوال اصلی که در این جا مطرح میشود این است که ایا این کافی است که توزیع هر متغیر تاثیر گذار را به صورت جداگانه در نظر بگیریم و دیگر به توزیع توامان آنها فکر نکنیم ؟ 
        پاسخ این است که در برخی موارد این کافی نیست. در ادامه بحث به حالاتی که میشود از توزیع توام صرف نظر کرد می‌پردازیم اما پیش از آن با یک مثال نقض به این سوال پاسخ میدهیم.
        بازی دارت را در نظر بگیرید که به حالت ساده شده ای درآمده است به این گونه که اگر درون دایره ای به شعاع R و مرکز دارت ضربه بزنید امتیاز کامل و در غیر اینصورت هیچ امتیازی نمی‌گیرید. بدیهتا اگر بخواهیم به توزیع امتیاز نسبت به متغیر تصادفی های  X و Y که به ترتیب مقدار طول و عرض جغرافیایی ضربه می‌باشند به صورت جداگانه توجه کنیم نمی‌توانیم پاسخ درستی به مساله دهیم و نیاز داریم از توزیع توام آنها کمک بگیریم.

<div id="sec_2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            توزیع توام گسسته و پیوسته
        </font>
		<p></p>
        <hr>
        برای تعریف توزیع توام بهتر است با تابع توزیع تجمعی یا  CDF شروع کنیم. با توجه به اینکه تعریف برای حالت پیوسته و گسسته صادق است در ادامه می‌توان از آن برای تعریف تابع چگالی احتمال در دامنه پیوسته و تابع جرم احتمال در دامنه گسسته کمک گرفت.
        <br>
        $$        
        F_{X,Y}(x,y) = P(X \le x \ ,\ Y \le y)
        \tag{1}
        $$
        <br>
        <span style="color:red">
   نکته قابل توجه در تعریف بالا این است که  X و Y میتوانند متغیر تصادفی هایی در  دامنه گسسته، پیوسته و یا یکی در گسسته و دیگری در پیوسته باشد.
        </span>
        <br>
   تعریف توام برای تابع جرم احتمال در دامنه گسسته به معنای اتفاق افتادن دو پدیده به صورت همزمان است بنابراین خواهیم داشت:
        <br>
        $$
        P_{X,Y}(x,y) = P(X=x \ , \ Y=y)
        \tag{2}
        $$
        <font color=#FF7500 size=6>
            مثال
        </font>
        <br>
        ظرفی شامل ۲ توپ سبز ۳ توپ آبی و ۴ توپ قرمز است. از این ظرف میخواهیم ۳ توپ را برداریم. متغیر تصادفی X      تعداد توپ های سبز و متغیر تصادفی  Y تعداد توپ های آبی در میان ۳ توپ انتخاب شده است. احتمال اینکه  X + Y
بزرگتر از ۱ شود چقدر است؟
        <br>
        $$ 
    P_{X,Y}(x + y > ۱) = P_{X,Y}(x + y = ۲) + P_{X,Y}(x + y = ۳) \\ = [ \, P_{x,Y}(x = ۰, y = ۲) + P_{X,Y}(x = ۱, y = ۱) + P_{X,Y}(x = ۲, y = ۰) ] \, \\ + [ \, P_{X,Y}(x = ۰, y = ۳) + P_{X,Y}(x = ۱, y = ۲) + P_{X,Y}(x = ۲, y = ۱) ] \,
    \\  = \dfrac{{۴ \choose ۱}{۳ \choose ۲}}{{۹ \choose ۳}} + \dfrac{{۲ \choose ۱}{۳ \choose ۱}{۴ \choose ۱} }{{۹ \choose ۳}} + \dfrac{{۲ \choose ۲}{۴ \choose ۱}}{{۹ \choose ۳}} + \dfrac{{۳ \choose ۳}}{{۹ \choose ۳}} + \dfrac{{۲ \choose ۱}{۳ \choose ۲}}{{۹ \choose ۳}} + 
    \dfrac{{۲ \choose ۲}{۳ \choose ۱}}{{۹ \choose ۳}} = \dfrac{۵۰}{۸۴}
        $$
        <br>
       در جدول پایین مقدار احتمال را به ازای تمامی مقادیر  X و  Y  می‌بینیم که همان تابع  pmf توام برای مثال بالا میباشد. 
         <figure>
            <img src="resources/pmf.png" , style="float:center">
            <center> <caption> تابع توزیع جرمی توام برای X, Y </caption> </center>            
        </figure>
        <br>        
        برای حالت پیوسته می‌خواهیم احتمال مقدار گرفتن متغیر تصادفی‌ها را در یک بازه پیوسته بیابیم که در حالتی که با یک متغیر تصادفی روبه‌رو بودیم این احتمال را با انتگرال گرفتن از تابع چگالی احتمال برای بازه مورد نظر بدست می‌آوردیم. حال در این فصل که با دو متغیر کار داریم، احتمال را با گرفتن انتگرال دوگانه روی آن بازه بدست می‌آوریم. برای درک بیشتر از ماهیت ریاضی موضوع تابع چگالی احتمال یک‌رویه است در فضای مختصاتی سه بعدی و احتمال آن بازه به معنای حجم محصور شده توسط بخشی از آن رویه یا همان بازه‌ی مذکور است. بدین ترتیب با استفاده از مطالب گفته شده داریم:
        <br>
         $$
        F_{X,Y}(x,y) = P(X \le x \ ,\ Y \le y) = \int_{-\infty}^{y} \int_{-\infty}^{x} f_{X,Y}(x,y) \ dx dy 
        \tag{3}
        $$
         <figure>
            <img src="resources/cdf.png" , style="float:center">
            <center> <caption> تابع توزیع تجمعی (F) روی بازه‌ی [3, 3-] </caption> </center>            
        </figure>
        <figure>
            <img src="resources/pdf.png">
            <center> <caption>  تابع چگالی احتمال  (f) روی بازه‌ی [3, 3-] </caption> </center>
        </figure>
        <br>
        حال بدیهتا با تعریف بالا داریم:
        $$
        F_{X,Y}({\infty},{\infty}) = 1       , F_{X,Y}({-\infty},{-\infty}) = 0
        $$
        <br> 
        همچنین با مشتق گیری جزئی از تعریف بالا تابع چگالی احتمال توام می‌شود:
        $$
        f_{X,Y} (x,y)= \frac{\partial ^ 2}{\partial x \partial y} F_{X,Y}(x,y)
        $$
        <br>
        در خیلی از مواقع نیاز است برای بدست آوردن احتمال در فضای پیوسته روی یک مجموعه انتگرال دو گانه بگیریم. اسم این مجموعه را C  میگذاریم و اگر داشته باشیم  $$C = \{(x, y) \mid x \in A , y \in B\}$$ 
        <br>
         آنگاه داریم:
         $$
         P_{X,Y}((x, y) \in C ) = \int_{B} \int_{A} f_{X,Y}(x,y) \ dx dy
         $$
          <font color=#FF7500 size=6>
            مثال
        </font>
        <br>
         X , Y دو متغیر تصادفی پیوسته هستند که تابع توزیع توام آن ها به صورت زیر می‌باشد .
         <br>
         $$
         f_{X,Y}(x, y) =
\left\{
	\begin{array}{ll}
		cx^2y  & \mbox{if } 0 \le y \le x \le 1 \\
		0 & \mbox{if } otherwise
	\end{array}
\right.
         $$
         <br>
         الف ) مقدار ثابت
         c را پیدا کنید .
         <br>
         ب ) $P(Y \le \dfrac{X}{2})$
         را محاسبه کنید.
         <br>
          الف ) 
         $$
         1 = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(x,y) \ dx dy \\
           = \int_{0}^{1} \int_{0}^{x} cx^2y \ dy dx \\
           = \int_{0}^{1} \dfrac{c}{2}x^4 dx \\
           = \dfrac{c}{10} \\
           \Rightarrow c = 10
         $$
         ب )
         $$
         P(Y \le \dfrac{X}{2}) = \int_{-\infty}^{\infty} \int_{0}^{\dfrac{x}{2}} f_{X,Y}(x,y) \ dx dy \\
         = \int_{0}^{1} \int_{0}^{\dfrac{x}{2}} 10x^2y \ dx dy \\
         = \int_{0}^{1} \dfrac{5}{4}x^4 dx \\
         = \dfrac{1}{4}
         $$
        <br>
        		<font color=#FF7500 size=6>
مثال
        </font>
        <br>
         می‌خواهیم توزیع متغیر تصادفی
        $ Z = X + Y $
        حساب کنیم.
        $$
        F_Z (z) = P(Z \leq z) = P( X + Y \leq z ) \\
                = \int \int_{X+Y\leq z} f_{X,Y}(x, y)
                = \int_{-\infty}^{\infty} \int_{-\infty}^{z-y} f_{X,Y}(x, y)\ dx dy
        $$
        و تابع چگالی احتمال هم با مشتق‌گیری از این تابع به دست می‌آید:
        $$
        f_Z(z) = \frac{d}{dz} F_Z(z) = \int_{-\infty}^{\infty}  f_{X,Y}(x, z-x)\ dx
        $$
        با همین روش میتوانیم توزیع هر تابعی از این دو متغیر را به دست آوریم

<div id="sec_3" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
توابع احتمال حاشیه‌ای
</font>
		<p></p>
        <br>
همانطور که مشاهده کردیداگر X و Y دومتغیر تصادفی گسسته باشند که روی یک فضای نمونه‌ای تعریف شده اند و مجموعه مقادیر ممکن Xو Y به ترتیب Aو B باشند؛ تابع 
       $$p(x,y) = P(X=x,Y=y)$$ 
       را تابع جرم احتمال توأم XوY می‌نامند. 
       <br>
       حال اگر X وY دارای تابع جرم احتمال 
       توأم 
       $p(x,y)$ 
       باشند و 
       $p_X$
       تابع جرم اجتمال X باشد؛ آنگاه داریم:
       $$p_X(x) = P(X=x)=P(X=x,y\subseteq B) = \sum_{y\subseteq B}P(X=x,Y=y) = \sum_{y\subseteq B}p(x,y)$$
       به همین صورت 
       $p_Y$ 
       تابع جرم احتمال Y می‌شود:
       $$p_Y(y) = \sum_{x\subseteq A}p(x,y)$$
<font color=blue size=6>
مثال:
</font>
        <br>
        دانشکده علوم ریاضی دانشگاهی دارای ۹۰ استاد مرد و ۳۰ استاد زن است. کمیته‌ای ۵ نفره به تصادف انتخاب می‌شوند تا هدفها و مسئولیتهای دانشکده را تهیه کنند. فرض کنید Xو Y به ترتیب تعداد مردان و زنان عضو کمیته باشد.
        <br>
        الف) تابع احتمال توأم XوY را تعیین کنید.
        <br>
        ب) $p_X$و $p_y$ ؛ یعنی توابع احتمال حاشیه‌ای XوY را بدست آورید.
        <br>
        <font color=blue size=5>
حل:
</font>
<br>
مجموعه مقادیر ممکن XوY هردو عبارت‌اند از 
$\{۰,۱,۲,۳,۴,۵\}$ 
تابع احتمال توأم XوY عبارت است از:
        $$p(x,y) = \frac{{۹۰\choose x}{۳۰\choose y}}{{۱۲۰\choose۵}} ,  x,y\subseteq\{۰,۱,۲,۳,۴,۵\}, x+y=۵$$
        و در غیر این حالت ۰ است.
        <br>
        و برای $p_x$و $p_Y$ داریم:
            $$p_X(x) = \sum_{y=۰}^{۵}p(x,y) = p(x,۵-x) = \frac{{۹۰\choose x}{۳۰\choose۵-x}}{{۱۲۰\choose۵}}  , \quad x\subseteq\{۰,۱,۲,۳,۴,۵\}$$
                    $$p_Y(y) = \sum_{x=۰}^{۵}p(x,y) = p(۵-y,y) = \frac{{۹۰\choose۵-y}{۳۰\choose y}}{{۱۲۰\choose۵}} ,  \quad y\subseteq\{۰,۱,۲,۳,۴,۵\}$$
                    <br>
                    در درسنامه های بعدی بیشتر درباره توزیع های حاشیه‌ای خواهیم دید.

<div id="sec_3" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
          توزیع توام برای بیش از 2 متغیر تصادفی  
        </font>
		<p></p>
        <br>
        تمامی فرمول هایی که برای ۲ متغیر تصادفی توام بررسی کردیم برای بیش از دو متغیر تصادفی نیز صدق میکند و به عنوان مثال داریم:
         $$
        F_{X_1,X_2,...,X_n}(x_1, x_2, ..., x_n) = P(X_1<x_1 \ , \ X_2<x_2 \ , ... ,\ X_n<x_n) \\
        f_{X_1,X_2,...,X_n}(x_1, x_2, ..., x_n) = \frac{\partial ^ n}{\partial x_1 \partial x_2 ... \partial x_n} F_{X_1,X_2,...,X_n}(x_1, x_2, ..., x_n)\\
        P(X_1 \in A_1, X_2 \in A_2, ... , X_n \in A_n) = \int_{A_n}\int_{A_n-1} ... \int_{A_1} f(x_1, x_2, ..., x_n) dx_1dx_2...dx_n   
        $$
        <font color=#FF7500 size=6>
            مثال
        </font>
        <br>
        فرض کنید ۹ بار یک تاس را پرتاب می‌کنیم که احتمال آمدن هر یک اعداد  ۱ تا ۶ در آن برابر است
           احتمال اینکه در این ۹ پرتاب ۲ بار یک، ۳ بار دو، ۱ بار سه، ۱ بار چهار و ۲ بار پنج بیاید چیست ؟
        <br>
هدف از طرح این مساله بیان مساله معروف‌تر توزیع چندجمله‌ای است. فرض کنید n پیشامد داریم که هرکدام r حالت برای رخ دادن دارند و احتمال پیش آمدن حالت $i$ برابر است با $p_i$($\sum_{i = 0}^{r}p_i = 1$) و 
همچنین تمام این پیشامد ها نسبت به هم مستقل هستند.
متغیر تصادفی $X_i$ را تعداد رخ دادن حالت  $i$ ام در این n 
پیشامد فرض میکنیم
و در اینصورت خواهیم داشت:
        $$
        P(X_1 = n_1, X_2 = n_2, X_3 = n_3, ... X_r = n_r) = \dfrac{n!}{n_1 ! n_2 ! ... n_r !}p_1^{n_1} p_2 ^{ n_2 }... p_r ^{ n_r}
        $$
           با شرط: $\sum_{i = 1}^{r}n_i = n$
        <br>
        توضیح ترکیبیاتی تعریف بالا به این صورت خواهد بود که فرض میکنیم $n_1$
          توپ از رنگ شماره 1 داریم
        $n_2$ توپ از رنگ شماره 2 و به همین صورت 
        $n_r$ توپ از رنگ شماره $r$ داریم
        که احتمال رخداد انها به ترتیب $p_1$ تا $p_r$ میباشد
        .بدست آوردن جایگشت این $n$ توپ یک مساله جایگشت با تکرار است که برابر میشود با $\dfrac{n!}{n_1 ! n_2 ! ... n_r !}$ و همچنین میدانیم هر جایگشت احتمالش برابر است با $p_1^{n_1} p_2 ^{ n_2 }... p_r ^{ n_r}$ . بنابراین به فرمولی همچون بالا میرسیم .
        همچنین جواب مساله برابر است با: 
        $$
        \dfrac{9!}{2!3!1!1!2!} (\dfrac{1}{6})^9
        $$

<div id="sec_4" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            توزیع توام بین چند تابع از چند متغیر تصادفی
        </font>
		<p></p>
		<hr>
        یاد گرفته‌ایم که از چند متغیر تصادفی توابعی بسازیم. اما ارتباط بین این توابع چگونه است؟ 
        <br>
        حالت ساده دو متغیره را در نظر بگیرید:
        <br>
        $$
        U = g_1(X,Y)
        $$
        $$
        V = g_2(X,Y)
        $$
        فرض کنید این تبدیلات وارون‌پذیر باشند و داشته باشیم:
        $$
        X = h_1(U,V)        
        $$
        $$
        Y = h_2(U,V)
        $$
        در این صورت می‌توان ثابت کرد:
        $$
        Jacoubian(u,v) = 
        \begin{bmatrix}
            \frac{\partial x}{\partial u} & \frac{\partial x}{\partial v} \\
            \frac{\partial y}{\partial u} & \frac{\partial y}{\partial v}
        \end{bmatrix}
        $$
        و خواهیم داشت:
        <br>
        $$
        x = h_1(u,v),
        y = h_2(u,v) \\
        \Rightarrow f_{U,V}(u,v) = |Jacoubian(u,v)|\ f_{X,Y}(x,y) 
        $$
         <font color=#FF7500 size=6>
مثال
        </font>
        <br>
        فرض کنید
        $X$
        و
        $Y$
        دو متغیر با توزیع نرمال توام گاوسی حول صفر با واریانس ۱ باشند.
        توزیع توام
        $
        َ U = \sqrt{X^2 + Y^2}
        $
        و
        $
        V = tg^{-1}(Y/X)
        $
        را به دست آورید.
	</font>
	</font>
</div>


<div id="sec_5" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            متغیر تصادفی های مستقل
        </font>
		<p></p>
        دو متغیر تصادفی $X$ و $Y$ مستقل از هم هستند اگر و تنها اگر به ازای هر زیرمجموعه دلخواه  $A$ و $B$ از اعداد حقیقی؛ پیشامدهای 
        $\{X\in A\}$و $\{Y\in B\}$ مستقل باشد یعنی اگر
        $$P(X \in A, Y \in B) = P(X \in A)P(Y \in B)$$
        تعریف بالا برای متغیر تصادفی‌های مستقل ما را به این می‌رساند که 
        $$
        P(X \le a, Y \le b) = P(X \le a)P(Y \le b)
        $$
         پس در نتیجه دو متغیر تصادفی مستقل هستند اگر و تنها اگر برای همه مقادیر حقیقی $a$ و $b$ داشته باشیم:
        $$
        F_{X,Y}(a, b) = F_{X}(a)F_Y(b)
        \tag{5}
        $$
        <br>
     این تعاریف برای دامنه گسسته به صورت $P(X = a, Y = b) = P(X = a)P(Y = b)$ و برای دامنه پیوسته به صورت $f_{X,Y}(a, b) = f_X(a)f_Y(b)$ در خواهد آمد 
     <br>
     برای مثال در تمرین احتمال در فوتبال وقتی می‌خواستیم احتمال شرطی ۳ گل زدن میزبان در صورتی که Real Madrid میزبان باشد را بدست بیاوریم؛ احتمال میزبان بودن رئال و ۳ گل زدن آن را از روی داده بدست می‌آوردیم عملا احتمال توام میزبان بودن رئال مادرید و۳ گل زدن را بدست می‌اوردیم. در ادامه بررسی خواهیم کرد آیا آن دو متغیرتصادفی مستقل بوده اند یا خیر؟ یعنی آیا احتمال توام میزبان بودن رئال مادرید و ۳ گل زدن برابر حاصل ضرب احتمال میزبانی رئال مادرید و احتمال ۳ گل زدن تیم میزبان است؟ 
     <br>
     فرض کنید XوY متغیرهای تصادفی مستقل گسسته به ترتیب با مجموعه مقادیر ممکن AوB باشند.در این صورت می‌توان نتیجه گرفت که برای همه مقادیر 
     $x\in A$و $y\in B$:
     $$P(X=x|Y=y)=P(X=x)$$
     و
     $$P(Y=y|X=x)=P(Y=y)$$
     پس Xو Y در صورتی مستقل‌اند که اطلاع از مقدار یکی از آن دو موجب تغییر تابع جرم احتمال دیگری نشود.
      <br>
      <font color=blue size=6>
مثال
        </font>
        <br>
        فرض کنید تعداد افرادی که در یک روز خاص به ایستگاه اتوبوس می آیند یک متغیر تصادفی poisson  با پارامتر $\lambda$ باشد . اگر $X$ متغیر تصادفی تعداد مردان و  $Y$ متغیر تصادفی تعداد زنان آمده به ایستگاه اتوبوس باشد و هم چنین احتمال آمدن مرد $p$ و احتمال آمدن زن $1-p$ باشد ،
        نشان دهید $X$ و $Y$ از هم مستقلند.
        $$
        P(X =i,Y =j)=P(X =i,Y =j \mid X + Y =i + j) P(X + Y =i + j)
        $$
        همچنین برای توزیع poisson داریم:
        $$
        P(X + Y = i + j) = \dfrac{e^{−λ} λ^{i+j}}{(i + j)!}
        $$
        همچنین بخش احتمال شرطی در معادله اول یک مساله برنولی است که با توجه به احتمال مرد و زن خواهیم داشت:
        $$
        P(X =i,Y =j \mid X + Y =i + j)= {{i +j} \choose i} p ^ i(1 − p)^j
        $$
         از کنار هم قرار دادن معادلات بالا میتوانیم تابع توزیع احتمال توام را بدست آوریم:
        $$
        P(X =i,Y = j)={i + j \choose i} p^i(1 − p)^j e^{−λ} \dfrac{λ^{i+j}} {(i + j)!} \\
        = e^{−λ} \dfrac{(λp)^i} {i!j!} (λ(1-p))^j \\
        =e^{λp}\dfrac{(λp)^i}{ i!} e^{−λ(1−p)}\dfrac{(λ(1 − p))^j}{ j!}
        $$
        حال میتوانیم $P(X = i)$ و $P(Y = j)$ را بدست آوریم تا بررسی کنیم که این دو متغیر تصادفی مستقل هستند یا نه 
        $$
        P(X = i) = e^{−λp}\dfrac{(λp)^i} {i!} \sum_j e^{−λ(1−p)}\dfrac{(λ(1 − p))^j} {j!}
        = e^{-λp}\dfrac{(λp)^i}{ i!}
        $$
        $$
        P(Y = j) = e^{-λ(1 - p)}\dfrac{(λ(1 - p)^j}{ j!} \sum_i e^{−λp}\dfrac{(λp)^i} {i!} = e^{-λ(1 - p)}\dfrac{(λ(1 - p)^j}{ j!}
        $$
         که نتیجه میشود $P(X = i, Y = j) = P(X = i)P(Y = j)$ و چون $i$ و $j$ مقادیر دلخواه بودند بنابراین تساوی بدست آمده به ازای همه مقادیر برقرار است و $X$ و $Y$ نسبت به هم مستقل هستند
