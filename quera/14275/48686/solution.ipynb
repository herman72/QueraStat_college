مقدمه
        </font>
		<p></p>
		<hr>
        در بسیاری از مسأله‌ها در آمار مقداری نامعلوم داریم که می‌خواهیم آن را تخمین بزنیم. برای مثال فرض کنید می‌خواهیم تعداد دانشجویان مرد را در دانشگاه تخمین بزنیم. (همچنین فرض کنید محاسبه‌ی این مقدار به‌طور مستقیم برایمان غیرممکن باشد) مرد بودن جنسیت هر دانشجوی رندومی از توزیع 
        $Bernoulli(p)$
   پیروی می‌کند که در آن 
   $p$
   مقداری ثابت است و برابر تعداد دانشجویان مرد به تعداد کل دانشجویان است. ۵ عدد از دانشجویان را با استفاده از شماره دانشجویی به صورت رندوم و مستقل از یکدیگر انتخاب می‌کنیم و مشاهده می‌کنیم که جنسیت هیچ کدام از آن‌ها مرد نیست. طبق روش‌هایی تا به حال برای تخمین استفاده می‌کردیم، مقدار 
   $p$
   برابر ۰ تخمین زده می‌شود. حال چرا نمی‌توانیم این مقدار تخمین‌زده‌شده را به عنوان مقدار واقعی قبول کنیم و نتیجه‌ی به‌دست‌آمده برایمان عجیب است؟ دلیل این امر این است که ما پیش‌فرض ذهنی‌ای داریم که تعداد دانشجویان مرد دانشگاه نزدیک به نصف کل دانشجویان آن است. پس روشی که بتواند این پیش‌فرض ما را به صورت اطلاعات وارد مسأله کند و از آن در تخمین استفاده کند مورد توجه‌مان خواهد بود.
روش‌های آماری که تا به اینجا مورد بحث قرار گرفته بودند همگی زیرمجموعه‌ رویکردی به نام فراوانی‌گرایی (frequentist) بودند. در مقابل این رویکرد، رویکرد دیگری به نام بیزی (bayesian) وجود دارد که در آن احتمال میزان باور ما را نشان می‌دهد. در این رویکرد می‌توانیم میزان باورمان را در مورد یک مقدار نامعلوم ثابت به صورت یک توزیع احتمالاتی نشان دهیم. در مثال گفته شده می‌توانیم فرض کنیم که مقدار نامعلوم 
$p$
از توزیع 
$Betta(2, 2)$
پیروی می‌کند که نشان‌دهنده‌ی باور ما نسبت به این مقدار نامعلوم است. در رویکرد فراوانی‌گرایی به علت ثابت بودن 
$p$
از چنین گزاره‌ای نمی‌توانستیم استفاده کنیم.
پیش از اینکه به معرفی روش بیزی بپردازیم مروری بر فلسفه‌ی بیزی خواهیم داشت. 
        </div>

<div id="bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مروری بر فلسفه بیزی
        </font>
		<p></p>
		<hr> 
        دید فراوانی‌گرایانه بر مبنای اصول زیر برقرار است:
        <ul>
        <li>
       احتمال یک اتفاق به معنای فرکانس نسبی آن نسبت به اتفاقات دیگر در تکرار‌های بسیار زیاد است. هم‌چنین احتمالات، به عنوان ویژگی‌های قابل مشاهده از جهان پیرامون شناخته می‌شوند.
       به معنای دیگر طرز تفکر یا دید ما نسبت به آن اتفاق تاثیری در احتمال نهایی آن ندارد و احتمال، ویژگی خود آن فرایند تصادفی یا اتفاق مربوط است.
       </li>
       <li>
       پارامتر‌های مسئله ثابت و البته ناشناخته هستند. برای مثال در دید فراوانی‌گرایانه فرض می‌کنیم داده‌ها از توزیع تصادفی با پارامتر ثابت و ناشناخته 
       $\theta$
   تولید می‌شوند. از آنجا که پارامتر‌ها نوسان ندارند و ثابت هستند، پس گزاره احتمالی به دردبخوری درباره آن‌ها نمی‌توان گفت.
       </li>
       <li>
       در دید فراوانی‌گرا رویکرد‌های آماری مناسبی بایستی طراحی شود که به ما در تکرار‌های زیاد کمک بکنند. برای مثال بازه اطمینان ۹۵ درصد بایستی مقدار واقعی پارامتر را در حداقل ۹۵ درصد از مواقع در بر بگیرد.
       </li>
       </ul>
       در مقابل رویکرد بیزی بر پایه این اصول بنا شده است:
       <ul>
       <li>
       احتمال، میزان باور ما را نشان می‌دهد. بنابراین ما می‌توانیم گزاره‌های احتمالی را در مورد چیز‌های وسیع‌تری به کار ببریم، نه فقط آن‌هایی که تغییرات رندوم دارند. برای مثال می‌توان گفت :«احتمال اینکه آلبرت انشتین در روز ۱ آگوست سال ۱۹۴۸ یک فنجان قهوه خورده باشد برابر ۰.۳۵ است.»
       این‌جا تکراری در اتفاق نداریم بلکه میزان باور خود به این اتفاق را به وسیله احتمال نشان دادیم.
       </li>
       <li>
       پس می‌توانیم در مورد پارامتر‌ها نیز گزاره‌های احتمالاتی بگوییم، هرچند مقادیر ثابتی باشند.
       </li>
       <li>
       استدلال‌های ما در مورد پارامتر 
       $\theta$
   با کمک تشکیل یک توزیع احتمالاتی برای 
       $\theta$
   نتیجه می‌دهند. بدین گونه می‌توان تخمین‌های نقطه‌ای و بازه‌ای را با دید جدیدی به دست آورد.
       </li>
       </ul>
       <br>
       هنوز هم جدال‌های فلسفلی بر سر دید بیزی وجود دارد. روش بیزی از آن‌جا که به ما اجازه دخیل کردن باور قبلی خود در مورد پارامتر را می‌دهد، اطمینانی در مورد جواب به دست آمده به ما نمی‌دهد. هم‌چنین ممکن است در تکرار‌های زیاد به جوابی متفاوت از جواب فراوانی‌گرایانه برسیم. اما امروزه استقبال زیادی از روش بیزی به خصوص در زمینه‌های استخراج داده (data mining) و یادگیری ماشین (machine learning) شده است.
       اما بحث‌های فلسفی به کنار بیاید ببینیم استنتاج بیزی چگونه انجام می‌شود!
        </div>

<div id="bayes2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
روش بیزی
</font>
		<p></p>
		<hr>
        در روش‌های بیزی ما در ابتدا یک پارامتر نامعلوم مانند 
        $\theta$
  داریم که با آن مانند یک متغیر تصادفی رفتار می‌کنیم. به‌طور دقیق‌تر، برای آن یک توزیع اولیه در نظر می‌گیریم که باور و حدس اولیه‌مان نسبت به توزیع این پارامتر است. به این توزیع prior گفته می‌شود.
  پس از مشاهده‌ی داده توزیع 
  $\theta$
   را بر اساس داده‌های مشاهده شده به‌روزرسانی می‌کنیم و به توزیع جدیدی می‌رسیم که به آن posterior گفته می‌شود. این مرحله معمولا با استفاده از قانون bayes انجام می‌شود.
      در ادامه مراحل استنتاج بیزی به صورت خلاصه آورده شده است:
      <ol>
      <li>
      یک تابع چگالی احتمال 
      $f(\theta)$
   را به عنوان توزیع prior انتخاب می‌کنیم.
      </li>
      <li>
      یک مدل آماری مناسب برای 
      $f(x|\theta)$
      انتخاب می‌کنیم. دقت کنید این‌بار به جای 
      $f(x;\theta)$
      می‌نویسیم 
      $f(x|\theta)$.
      </li>
      <li>
      بعد از مشاهده داده‌ها باور خود را مورد پارامتر بروز‌رسانی می‌کنیم و 
      $f(\theta | x)$
      را محاسبه می‌کنیم.
      </li>
      </ol>
   برای درک مرحله سوم ابتدا فرض کنید 
      $\theta$
      از توزیع گسسته‌ای آمده باشد و تنها یک داده گسسته x داریم.
      طبق قانون بیز به دست می‌آید:
      $$ P(\Theta = \theta | X = x) = \frac{P(X = x , \Theta = \theta)}{P(X = x)} = \frac{P(X=x|\Theta=\theta)P(\Theta=\theta)}{\sum_\theta P(X=x|\Theta = \theta)P(\Theta = \theta)} $$
      برای حالت پیوسته نیز با جاگذاری تابع چگالی احتمال به دست می‌آوریم:
      $$f(\theta|x) = \frac{f(x|\theta)f(\theta)}{\int f(x|\theta)f(\theta)d\theta}$$
        حال اگر n داده IID داشته‌باشیم، داریم:
        $$f(x_1, ..., x_n | \theta) = \Pi_{i=1}^n f(x_i | \theta) = L(\theta | X)$$
        فرض کنید $x^n = (x_1, ..., x_n)$ آنگاه
        $$f(\theta|x^n)=\frac{f(x^n|\theta)f(\theta)}{\int f(x^n|\theta)f(\theta)d\theta} = \frac{L(\theta|X)f(\theta)}{c_n} \propto L(\theta|X)f(\theta)$$
        که در آن
        $$c_n = \int L(\theta | X)f(\theta)d\theta$$
        که 
        $L(\theta|X)$
        تابع likelihood و $c_n$
        ثابت نرمال‌سازی (normalizing constant) نامیده می‌شود.
        دقت کنید که 
        $c_n$
        به $\theta$
        وابسته نیست.
        ممکن است فکر کنید دور‌انداختن ثابت $c_n$
            برای‌ما مشکل ایجاد کند اما می‌توانیم در موقع نیاز این ثابت را باز‌یابی کنیم. حتی ممکن است اگر posterior ما از یکی از توزیع‌های معروف پیروی کند بدون محاسبه انتگرال، ثابت را به دست آورد!
            <br>
            حال چه استفاده‌ای از توزیع posterior می‌توان کرد؟
            می‌توانیم مٌد (mode) یا میانگین توزیع posterior را به دست آوریم.
            هم‌چنین می‌توان یک تخمین بازه‌ای بیزی به دست آورد. مثلا a و b ای را به دست می‌آوریم که :
           $$\int_{-\infty}^{a}f(\theta|X)d\theta = \int_{b}^{\infty}f(\theta|X)d\theta=\frac{\alpha}{2}$$ 
           و اگر قرار دهیم 
           $C=(a,b)$
           آنگآه 
           $$P(\Theta \in C|X) = \int_a^b f(\theta|x)d\theta= 1-\alpha$$
           پس C یک تخمین 
           بازه‌ای بیزی با احتمال $1-\alpha$
           می‌باشد.
        </div>

<div id="MAP" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
روش MAP
</font>
		<p></p>
		<hr>
        اگر تخمین ما بر اساس داده‌ها مقدار مد توزیع  posterior باشد آنگاه نوع تخمین‌گر ما  
      maximum a posteriori estimation
      یا به اختصار
      MAP
       نامیده می‌شود.
       بنابراین:
       $$ \hat{\theta}_{MAP}(x) = \underset{\theta}{argmax}\ f(\theta|X) = \underset{\theta}{argmax}\ \frac{f(X|\theta)f(\theta)}{c_n} = \underset{\theta}{argmax}\ f(X|\theta)f(\theta)$$
       بگذارید با یک مثال پیش برویم. فرض کنید داده‌های ما از توزیع هندسی با پارامتر $\theta$ می‌آیند.
       هم‌چنین پیش فرض ما این است که توزیع پارامتر بدین شکل است:
       $$
 \nonumber f_\theta(\theta) = \left\{
  \begin{array}{l l}
    2\theta  &  \quad  \textrm{if }0 \leq \theta \leq 1  \\
      &  \quad   \\
    0 &  \quad \text{otherwise}
  \end{array} \right.
       $$
       اولین داده‌ای که به دست ما می‌رسد x=3 است و می‌خواهیم آن را به روش MAP تخمین بزنیم. فهمیدیم که $\hat{\theta}_{MAP}$ برابر پارامتری است که مقدار 
       $\underset{\theta}{argmax}\ f(x|\theta)f(\theta)$ را بیشینه کند.
       پس داریم:
       $$
       P(x=3|\theta) = \theta * (1-\theta)^2 \\
       \Rightarrow P(\theta | x=3 ) \propto \theta * (1-\theta)^2 * f(\theta) \\
       $$
       برای بیشنیه کردن این عبارت از آن نسبت به $\theta$ مشتق می‌گیریم و آن را برابر صفر قرار می‌دهیم.
       $$
    \frac{\textrm{d}}{\textrm{d}\theta} \bigg[\theta^2(1-\theta)^2 \bigg]=2\theta(1-\theta)^2-2(1-\theta)\theta^2=0.
    \Rightarrow \hat{\theta}_{MAP} = \frac{1}{2}
$$
<br>
حال به مثالی که در مقدمه مطرح کردیم می‌پردازیم. در آن مثال هدفمان تخمین پارامتر $p$ بود. همان‌طور که دیدیم نتیجه‌ی تخمین‌گر ML به واقعیت نزدیک نبود. به همین دلیل از رویکرد بیزی استفاده کردیم و پیش‌فرضمان این بود که مقدار $p$ نزدیک به ۰.۵ است. حال به توزیعی برای prior نیاز داریم که این پیش‌فرض را به خوبی نشان دهد. طبق شکل زیر توزیع 
$Betta(2, 2)$
کاندید بسیار خوبی برای prior است.
        </div>

```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

x = np.arange(0, 1, 0.001)
beta_pdf = scipy.stats.beta.pdf(x, 2, 2)

plt.plot(x, beta_pdf)
plt.ylabel("beta(2, 2)")
plt.show()
```

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
         این توزیع را به عنوان prior انتخاب می‌کنیم. حال با استفاده از ۵ داده‌ای که در دست داریم باید توزیع posterior را از روی prior به دست آوریم. دقت کنید که داده‌هایمان همگی ۰ بودند. داریم:
        $$f(\theta | x_1, ..., x_5) \propto L(\theta | X)f(\theta) = \Pi_{i=1}^{5}f(x_i|\theta) \ f(\theta)$$
   از آن‌جایی که 
   $f(x_i|\theta) = \theta ^{x_i} (1-\theta)^{1-x_i}$
   و
   $f(\theta) = \theta (1-\theta)$
   داریم:
   $$f(\theta | x_1, ..., x_5) \propto \theta ^{\sum x_i} (1-\theta)^{5 - \sum x_i} f(\theta) = \theta ^{1 + \sum x_i} (1-\theta)^{6 - \sum x_i} = \theta (1-\theta)^6$$
   می‌دانیم که 
   $Betta(2, 7) \propto \theta (1 - \theta)^6$
   .
   پس توزیع posterior همان 
   $Betta(2, 7)$ 
   است.
   در شکل زیر تابع likelihood و توزیع posterior نشان داده شده‌اند.
</font>
</div>

```python
import matplotlib.pyplot as plt
import scipy.stats, scipy.special
import numpy as np

likelihood = scipy.stats.binom.pmf(0, 5, x)

plt.subplot(1, 2, 1)
plt.plot(x, likelihood)
plt.title("Bernouli likelihood function")

plt.subplot(1, 2, 2)
plt.plot(x, np.multiply(beta_pdf, likelihood) * 
         scipy.special.beta(2, 2) / scipy.special.beta(2, 7))  # beta functions are used as normalizing constant
plt.title("Posterior")
plt.show()
```

	<font face="XB Zar" size=5>
         همان‌طور که در شکل نیز دیده می‌شود مد توزیع posterior برابر 
        $\frac{1}{7}$
   است که در 
   $\frac{d}{d\theta}[\theta (1-\theta)^6] = 0$
   صدق می‌کند. پس در رویکرد بیزین تخمین ما از 
   $\theta$
   برابر $\frac{1}{7}$ خواهد بود.
</font>
</div>

انتخاب Prior
</font>
		<p></p>
		<hr> 
سوالی که اکنون پیش می‌آید این است که  توزیع prior یک پارامتر را از کجا بایستی فهمید؟ و اینکه چگونه بفهمیم که prior 
ما خوب است یا بد؟
جواب به این دو سوال سخت است! جواب تقریبی به این سوال این است که توزیع prior را متخصصان حوزه مربوط به آن مسئله انتخاب می‌کنند. انتخاب prior مناسب در بعضی مواقع نیاز به دانش تخصصی خاص در آن حوزه دارد.
هم‌چنین در این مورد ابزاری از دنیای ریاضیات هم به کمک آمار‌دان‌ها آمده است! برای توزیع‌های خاصی به عنوان $f(X|\theta)$
توزیع‌های prior انتخاب می‌شوند که به ما در ساده سازی محاسبات کمک بسیاری می‌کنند.
<br>
برای مثال فرض کنید که توزیع 
$f(x|\theta)$
را برنولی در نظر گرفته‌ایم. توزیع prior‌ای که معمولا برای برنولی در نظر می‌گیرند، توزیع بتا می‌باشد. 
        </div>

<div id="pros" style="direction:rtl;line-height:300%;">

ضعف‌ها و قوت‌های رویکرد بیزی
</font>
		<p></p>
		<hr>
        برخی رویکرد بیزی را جذاب می‌بینند زیرا با استفاده از آن می‌توانیم گزاره‌های احتمالی در مورد پارامتر بگوییم. برای مثال فرض کنید سکه‌ای داریم که می‌خواهیم احتمال رو آمدن آن را دریابیم (یا ببینیم چقدر سکه سالم است).
        فرض کنید هیچ پیش فرضی در مورد این احتمال نداریم. پس توزیع پارامتر $\theta$ را 
        که احتمال رو آمدن سکه است را
        Uniform
        از ۰ تا ۱ 
        در نظر می‌گیریم.
       سکه را پرتاب می‌کنیم و سکه هر بار «رو» می‌آید. می‌خواهیم اثر این داده را بر روی توزیع posterior خود بررسی کنیم.
       متغیر n را تعداد یک آمدن‌های سکه در نظر بگیرید.
      توزیع‌های نهایی بدین شکل خواهند بود:
        </div>

```python
from scipy.stats import uniform, binom
import matplotlib.pyplot as plt

f1 = lambda x: binom.pmf(1, 1, x) * 2 * uniform.pdf(x)

f2 = lambda x: binom.pmf(2, 2, x) * 3 * uniform.pdf(x)

f5 = lambda x: binom.pmf(5, 5, x) * 6 * uniform.pdf(x)

f8 = lambda x: binom.pmf(8, 8, x) * 9 * uniform.pdf(x)

x = np.arange(0, 1, 0.001)
plt.plot(x, uniform.pdf(x), label="prior")
plt.plot(x, f1(x), label="n=1")
plt.plot(x, f2(x), label="n=2")
plt.plot(x, f5(x), label="n=5")
plt.plot(x, f8(x), label="n=8")
plt.legend()
plt.show()
```

		<p></p>
		<hr>
        همان طور که می‌بینید در توزیع posterior 
        احتمال $\theta$
        های کوچک برای n‌های بزرگ، کوچک و کوچک تر می‌شود.
        اما رویکرد بیزی احتمال را برای این $\theta$‌ها صفر نمی‌کند و این عدم قطعیت کمک بزرگی است. همچنین با بزرگ‌تر شدن n احتمال یک بودن پارامتر بزرگ و بزرگ‌تر می‌شود. علاوه بر این با افزایش n واریانس توزیع نیز کاهش می‌یابد و درباره‌ی تخمینی که در این رویکرد گزارش می‌کنیم بیشتر مطمئن می‌شویم. در مدل‌های پارامتری با تعداد داده‌های زیاد (یعنی بزرگ‌تر شدن n)
        رویکرد بیزی و فراوانی‌گرایانه نتایج تقریبا یکسانی به ما خواهند داد. اما در حالت کلی لزوما یکسان نخواهند شد. (می‌توانید برای این حالت مثال بزنید؟)
        <br>
        نکته مهمی که از نمودار می‌توان فهمید این است که رویکرد بیزی به ما 
        $P(\theta \in [a,b]|X)$ را می‌دهد. یعنی با مشاهده داده‌ها و تشکیل prior می‌توان گفت پارامتر ما با چه احتمالی در هر بازه دلخواه است. دقت کنید که روش فراوانی‌گرایانه به ما این امکان را نمی‌داد و تنها می‌توانستیم در مورد $\theta$
        بازه اطمینان به دست آوریم. بگذارید با مثالی این موضوع را روشن‌تر کنم:
        <br>
        فرض کنید $\theta$
        عددی حقیقی و ثابت باشد و $X_1$ و $X_2$ دو متغیر تصادفی مستقل باشند به طوری که 
        $P(X_i = 1) = P(X_i = -1) = \frac{1}{2}$.
        حال تعریف می‌کنیم 
        $Y_i = \theta + X_i$
        و فرض کنید تنها $Y_1$ و $Y_2$ را مشاهده می‌کنیم.
        تعریف می‌کنیم:
        $$ C = {Y_1 -1} \ if:\ Y_1= Y_2 \\
        C = {\frac{(Y_1 + Y_2)}{2}} \ if: \ Y_1 \neq Y_2 $$
            C یک بازه احتمال ۷۵ درصدی است زیرا $\theta$ هر چه باشد، $P_\theta(\theta \in C) = 3/4$. حال فرض کنید مشاهده می‌کنیم که $Y_1 = 15$ و $Y_2 = 17$ که در اینصورت $C = \{16\}$. با این حال در این مورد مطمئنیم که $\theta = 16$. گفتن این که C یک بازه اطمینان ۷۵ درصدی است بعضی را گیج می‌کند با این حال این بازه در ۷۵ درصد مواقع مقدار واقعی $\theta$ را در خود دارد. 
            در این مورد رویکرد بیزی برای مردم ملموس‌تر است. برای سادگی فرض کنید $\theta$ عدد صحیح باشد و $f(\theta)>0$ برای هر عدد صحیح. برای 
            $Y=(Y_1, Y_2) = (15,17)$ تابع likelihood برابر است با:
            $$ L(\theta) = 1/4 \ if\ \theta = 16 \\
            L(\theta) = 0 \ otherwise$$
            با اعمال قانون بیز به دست می‌آید:
            $$ P(\Theta = \theta | Y= (15,17)) = 1 \ if\ \theta = 16
            \\
            P(\Theta = \theta | Y = (15,17)) = 0 \ otherwise$$
            بنابراین $P(\theta \in C|Y = (15,17)) = 1)$
            هیچ اشکالی در گفتن این‌که {16} یک بازه اطمینان ۷۵ درصدی است، نیست اما این یک جمله احتمالی در مورد $\theta$ نیست.
            <br>
            اما یک ضعف اصلی رویکرد بیزی این است که می‌تواند نتایج کاملا گمراه کننده و غلطی به ما بدهد که حساس بودن انتخاب prior را به ما گوشزد می‌کند! همان‌طور که قبلا گفتیم در تکرار‌های زیاد انتظار داریم تخمین‌گر بیزی همان نتیجه تخمین‌گر ML را بدهد اما انتخاب یک prior نامناسب می‌تواند ما را از جواب واقعی دور کند.
        </div>
