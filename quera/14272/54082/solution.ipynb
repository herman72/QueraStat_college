		<p></p>

در بخش قبل با امید ریاضی آشنا شدیم و دیدیم که به کمک این کمیت مهم می‌توان فهمید که نتیجهٔ آزمایش ها حدودا حول چه مقداری است.        
<br/>
در این قسمت کمیت دیگری به نام واریانس را معرفی می‌کنیم. امید ریاضی با وجود این‌که در مورد مقدار میانگین اطلاع می‌داد، در مورد

<div id="sec_def_var" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف واریانس
        </font>
		<p></p>
		<hr>
       واریانس یک متغیر تصادفی گسسته
را که معیاری از پراکندگی داده خروجی‌های ممکن آن است،
به شرح زیر تعریف می‌کنیم:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mathop{\mathbb{E}[X]})^2]
$$
یا در برخی نمادگذاری‌های دیگر به صورت زیر تعریف می‌شود:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mu)^2]
$$
که در آن
$\mathop{\mathbb{E}}[X] = \mu$
است.
<br/>
با ساده کردن عبارت بالا داریم:
$$
\begin{aligned}
Var(X) &= \mathop{\mathbb{E}}[(X - \mu)^2] \\
&= \sum_{x}(x-\mu)^2p(x) \\
&= \sum_{x}(x^2+\mu^2-2\mu{x})p(x) \\
&= \sum_{x}x^2p(x) + \sum_{x}\mu^2p(x) -2\mu\sum_{x}xp(x)\\
&= \mathop{\mathbb{E}}[X^2] + \mu^2 -2\mu.\mu \\
&= \mathop{\mathbb{E}}[X^2] - \mu^2 \\
Var(X) &= \mathop{\mathbb{E}}[X^2] -  (\mathop{\mathbb{E}}[X])^2\\
\end{aligned}
$$
<br/>
        هم چنین واریانس را با 
        $\sigma^2$
        نیز نشان می دهند.
<br/>
همچنین به جذر واریانس، انحراف معیار می‌گویند.
$$
SD(X) = \sqrt{Var(X)} = \sigma
$$
در برخی جاها واریانس دید بهتری می‌تواند به ما بدهد تا انحراف معیار و در برخی جاهای دیگر برعکس، به همین دلیل با اینکه با هم رابطه ی بسیار ساده‌ای دارند و با داشتن یکی دیگری را به سادگی می‌توان به دست آورد، هردوی آن ها استفاده می‌شوند.
    </font>
</div>

<div id="sec_prop_var" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص واریانس
        </font>
		<p></p>
		<hr>
واریانس معیاری از پراکندگی خروجی‌های مختلف یک متغیر تصادفی است.
بنابراین انتقال یک متغیر تصادفی به اندازه یک عدد ثابت نباید اثری در آن داشته باشد(یعنی اگر تمام اعدادی که متغیر تصادفی نسبت می‌دهد را به علاوهٔ یک عدد ثابت کنیم) چون پراکندگی نسبی خروجی‌های متغیر تصادفی انتقال‌یافته، نسبت به یکدیگر عوض نمی‌شود.
<br/>
همچنین واریانس با توان دوم مرتبط است. بنابراین دور از  
انتظار نیست که ضرب یک متغیر تصادفی در مقدار
ثابت رابطه توان دوم با واریانس متغیر اولیه داشته باشد.
<br/>
دو خاصیت بالا را در رابطه زیر می‌یابیم:
$$
\begin{aligned}
Var(aX + b) &= \mathop{\mathbb{E}}[(aX+b-a\mu-b)^2] \\
&= \mathop{\mathbb{E}}[a^2(X-\mu)^2] \\
&= a^2\mathop{\mathbb{E}}[(X-\mu)^2] \\
&= a^2Var(X)
\end{aligned}
$$
    </font>
</div>

مثال
        </font>
		<p></p>
		<hr>
در جعبه‌ای
تعداد ۵ گوی آبی و ۵ گوی قرمز است.
دو گوی را به تصادف بدون جایگذاری از جعبه در می‌آوریم.
اگر دو گوی هم‌رنگ باشند
۱.۱
دلار
به دست می‌آوریم و اگر ناهم‌رنگ باشند ۱ دلار می‌بازیم.
اگر متغیر تصادفی مقدار پولی که
به دست می‌آوریم
،
$X$
باشد،
$Var(X)$
چند است؟
<br/>
$$
\mathop{\mathbb{E}}[X] = +1.1 \times (\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) + (-1) \times (1-\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) = -\frac{1}{15}
$$
$$
\mathop{\mathbb{E}}[X^2] = +1.1^2 \times (\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) + (-1)^2 \times (1-\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) = \frac{82}{75}
$$
$$
\begin{aligned}
Var(X) &= \mathop{\mathbb{E}}[X^2] - \mathop{\mathbb{E}}[X]^2\\
&= \frac{82}{75} - (\frac{-1}{15})^2\\
&= 1.0889 \\
\end{aligned}
$$
    </font>
</div>

<div id="sec_bernoulli" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه        </font>
		<p></p>
		<hr>
        تا کنون با متغیر های تصادفی آشنا شدیم و فهمیدیم با کمک امید ریاضی و واریانس می‌توانیم شهودی از این که داده های ما حول چه عددی هستند(امید ریاضی) و این که تا چه حد به این میانگین نزدیک هستند(واریانس) بدست آوریم. حال به این می‌پردازیم که هر عددی که متغیر تصادفی عضوی را به آن می‌نگارد، چه احتمالی دارد. در واقع اگر نمودار 
        $F(x) - x$
        را رسم کنیم، از چه الگویی پیروی می‌کند.
        <br/>
        به شکلی که متغیر تصادفی خواهد داشت، توزیع می‌گویند. می‌توان این توزیع را با $F(x) - x$ 
        نشان داد یا بر حسب چیزهایی که در آینده با آن ها آشنا خواهیم‌شد. در توزیع های مختلفی که وجود دارد، توزیع هایی هستند که متداول‌تر هستند. در این بخش می‌خواهیم با تعدادی از این توزیع‌ها آشنا شویم. وقتی یک متغیر تصادفی مثل $X$ 
        یک توزیع خاص داشته‌ باشد(مثلا توزیع برنولی که به زودی با آن آشنا می‌شویم) می‌گوییم متغیر تصادفی $X$ 
        از آن توزیع پیروی می‌کند. هم چنین هر توزیعی یک یا چند پارامتر دارد که بر حسب آن‌ها، شکل دقیقشان مشخص می‌شود(پارامتر یا عدد هستند یا خودشان یک متغیر تصادفی هستند) و وقتی گفته می‌شود متغیر تصادفی از یک توزیع پیروی می‌کند باید پارامتر های آن توزیع هم مشخص شود.
        <br/>
        ابتدا با تعدادی از توزیع‌های گسسته آشنا می‌شویم و بعد از آن به سراغ متغیر‌های تصادفی پیوسته می‌رویم. برای هر توزیع هم امیدریاضی، واریانس و برخی دیگر از ویژگی‌هایشان را محاسبه می‌کنیم.
        </font>

<div id="sec_bernoulli" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
توزیع گسسته برنولی
        </font>
		<p></p>
		<hr>
        توزیع برنولی، توزیعی گسسته است که نام آن از نام دانشمند سوئیسی ژاکوب برنولی گرفته شده‌است. توزیع برنولی، توزیع مربوط به انجام یک آزمایش است که نتیجهٔ آن موفقیت یا عدم موفقیت در آزمایش است. توزیع برنولی در صورت موفقیت به نتیجهٔ آزمایش، عدد 1 را نسبت می‌دهد و اگر آزمایش موفقیت آمیز نباشد، عدد 0 را نسبت می‌دهد.
        <br/>
        احتمال موفقیت آزمایش برابر p است و احتمال شکست آن برابر q=1-p است.به p پارامتر توزیع گفته می‌شود.
        اگر متغیر تصادفی $X$
        از توزیع برنولی با پارامتر p پیروی کند می گوییم
        $X \sim Bernoulli(p)$.
        در این صورت داریم :
        \[
        Pr(X=1)=p=1-Pr(X=0)=1-q
        \]<br>
            تصویری از توزیع برنولی :

        </li>

امیدریاضی و واریانس توزیع

        <ul>
            <li>
                امیدریاضی:
                <br>
                \begin{aligned} \operatorname {E}[X]=p \end{aligned}
                <br>
            $$E[x] = Pr(X=1) \times 1 + Pr(X=0) \times 0 = p \times 1 + q \times 0 = p$$
                <br>
            </li>
            <li>
                واریانس:
                <br>
                 \begin{aligned} \operatorname {Var} (X)=p - p^2
                 \end{aligned}

            </li>

<div id="sec_expectation_binomial" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            مثال توزیع برنولی
        </font>
		<p></p>
		<hr>
		<p></p>
            نوع دیگری که مسئله‌های برنولی مطرح می‌شوند ترکیبی از تکرار یک کار است که هر بار آن فقط موفقیت و یا شکست دارد. برای مثال مسئله زیر را در نظر بگیرید:
    <br>
    یک سکه را ۵ بار پرتاب می‌کنیم به چه احتمالی (شیر، خط، شیر، شیر، خط) به ترتیب از راست به چپ ظاهر می‌شود.
    <br>
    <br>
        در این مسئله در پرتاب اول شیر آمدن موفقیت است و در پرتاب دوم خط آمدن و به همین ترتیب. اما در همه پرتاب‌ها امکان موفقیت $\frac{1}{2}$ و امکان شکست نیز $\frac{1}{2}$ است.
    <br>
    $$Pr = \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} = \frac{1}{2 ^ 5}$$
        <br>
    </font>

<div id="sec_problem_binomial" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            توزیع گسسته دو جمله‌ای
        </font>
		<p></p>
		<hr>
        این توزیع مربوط به موفقیت تعدادی از آزمایش هایی است که پشت سر هم انجام می شوند و از هم مستقل هستند(انگار که
        $n$
        آزمایش برنولی پشت هم انجام شده است). حال می‌خواهیم بررسی کنیم که اگر از $n$ 
        آزمایش که هر کدام به احتمال $p$ موفقیت آمیز است،
        به چه احتمالی $k$ آزمایش
        موفقیت آمیز بوده‌اند ؟
        <br/>
        فرض کنید که $n$ آزمایش را انجام داده‌ایم.
        متغیر تصادفی $X$ تعداد آزمایش‌های موفق را نشان می‌دهد.
        می‌خواهیم 
        $Pr(X = k)$
        را بدست آوریم.
        در این صورت می‌گوییم 
        $X \sim Binomial(n, p)$.
        برای محاسبه احتمال هم باید از $n$ آزمایش،
        دقیقا $k$  تا را موفق شویم
        و $n - k$ 
        تا را مردود شویم. هم چنین می‌دانیم احتمال موفقیت هر آزمایش مستقل از سایر آزمایش‌ها، $p$
        و احتمال مردود شدن هم به صورت مستقل از سایر آزمایش‌ها 
        $1 - p$
        است. اگر بدانیم کدام آزمایش‌ها را موفق می‌شویم و کدام‌ها را مردود می‌شویم، احتمال این نتیجه برابر است با:
        <br/>
        $$Pr = p^k(1-p)^{n-k}$$
        <br/>
                با توجه به فرمول بالا، می‌فهمیم که ترتیب آزمایش‌هایی که قرار است موفقیت آمیز باشند مهم نیست و 
         کافی است تعداد حالات مختلف $k$ بار موفقیت را در $n$ آزمایش حساب کنیم که برابر است با:
        ${n \choose k}$
        <br>
        و فرمول کلی می‌شود:   
        $$P(k)={n \choose k}p^{k}(1-p)^{n-k}$$
        <br>
        اعداد
         ${n \choose k}$
         ضرایب دوجمله‌ای نامیده می شوند.
        <br>
        متغیر تصادفی $X$ که  به صورت زیر تعریف شود از توزیع دو جمله‌ای پیروی می‌کند: 
        <br>
        $$ X = occurrences\: of\: event\: A\: in\: n \: trials $$
        <br>
        $$X \sim binomial(n,p)$$
        <br>
        توزیع دوجمله‌ای با n = 100 و p = 0.5 :
         <img src="figs/binomial.png" style="float:center; width:70%"/>
    </font>
</div>

امیدریاضی و واریانس توزیع
            دوجمله‌ای
        </font>
		<p></p>
		<hr>
        <ul>
            <li>
                امیدریاضی:
                <br>
                \begin{aligned} \operatorname {E}[X]=np \end{aligned}
                <br>
         \begin{aligned}\mu &=\sum _{k=0}^{n}k{\binom {n}{k}}p^{k}(1-p)^{n-k}\\&=np\sum _{k=0}^{n}k{\frac {(n-1)!}{(n-k)!k!}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\&=np\sum _{k=1}^{n}{\frac {(n-1)!}{((n-1)-(k-1))!(k-1)!}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\&=np\sum _{k=1}^{n}{\binom {n-1}{k-1}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\&=np\sum _{\ell =0}^{n-1}{\binom {n-1}{\ell }}p^{\ell }(1-p)^{(n-1)-\ell }&&{\text{with }}\ell :=k-1\\&=np\sum _{\ell =0}^{m}{\binom {m}{\ell }}p^{\ell }(1-p)^{m-\ell }&&{\text{with }}m:=n-1\\&=np(p+(1-p))^{m}\\&=np\end{aligned}
            بعدا که با خواص امیدریاضی آشنا شدیم، می‌توانیم ساده‌تر نیز این امید ریاضی را حساب کنیم.
            </li>
            <li>
                واریانس:
                <br>
                 \begin{aligned} \operatorname {Var} (X)=np(1-p)
                 \end{aligned}
                در آینده با ابزار محاسبهٔ این واریانس آشنا می‌شوید.
            </li>
        </ul>
    </font>

سوالات هیچ ربطی به هم ندارند و کاملا از هم مستقل اند.
سوالات چهارگزینه‌ای هستند و شرکت کننده ها کاملا تصادفی به سوالات پاسخ می دهند. حال میخواهیم بدانیم به صورت میانگین، چه میزان باید برای برگزاری مسابقه بین 10 نفر جایزه بدهیم.
<br>

مساله دو توزیع دوجمله‌ای است. در هر سوال با فرض اینکه شخص هیچ دانشی نسبت به سوال ندارد، و اینکه ۴ گزینه وجود دارد، احتمال موفقیت ۰.۲۵ است.توزیع اول برای هر فرد و سوالاتش برقرار است. یعنی ۲۰ بار برای هر فرد آزمایش یکسانی تکرار می شود که احتمال موفقیت ۰.۲۵ است. حالا در ۲۰ سوال احتمال اینکه یک نفر بیشتر مساوی ۱۸ سوال پاسخ درست بدهد برابر است  با احتمال اینکه: دقیقا ۱۸ سوال + دقیقا ۱۹ + دقیقا ۲۰ سوال پاسخ درست بدهد.  
<br>
$$p(success\: for\: an\: individual) = p(18) + p(19) + p(20) ={20 \choose 18}(0.25)^{18}(1-0.25)^{20-18} + {20 \choose 19}(0.25)^{19}(1-0.25)^{20-19} + {20 \choose 20}(0.25)^{20}(1-0.25)^{20-20} = 1.55524 \times 10^{-9} + 5.457 \times 10^{-11} + 9.1 \times 10^{-13} = 1.61072 \times 10^{-9} $$
<br>
        در مرحله بعد ۱۰ نفر داریم که هر نفر احتمال موفقیتی برابر مقدار بدست آمده در رابطه بالا دارد. بطور متوسط چقدر باید پول بدهد؟  باید امید ریاضی  توزیع جدید را بیابیم.
<br>
$$X \sim binomial(n=10,p=1.61072 \times 10^{-9})$$
<br>
$$E = np = 1.61072 \times 10^{-8}$$
