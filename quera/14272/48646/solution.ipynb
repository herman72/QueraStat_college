<div id="sec_intro_rv" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در قسمت‌های قبلی درس یاد گرفتیم که فضای نمونه
مانند
$\Omega$،
مجموعه‌ای است که شامل تمام خروجی‌های مختلف یک آزمایش تصادفی است. اگر بخواهیم نتیجه ی یک آزمایش را بررسی کنیم، کار کردن با اشیایی که عدد نیستند کار ما را سخت می‌کنند؛ زیرا عملگر‌های زیادی روی آنها نداریم. در نتیجه به دنبال مبدلی هستیم که اعضای 
        $\Omega$
        را به اعداد تبدیل کند. 
        متغیرهای تصادفی توابعی هستند که این خروجی‌ها را به اعداد حقیقی نگاشت می‌کنند.
<br/>
</div>

<div id="sec_def_rv" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف
            <p></p>
        </font>
        <hr>
        به یک تابع مانند
        $X$
        که اعضای 
        $\Omega$
        را به 
        $R$
        نگاشت می‌کند، یک متغیر تصادفی می‌گویند.
        در نتیجه به ازای هر عضو $\Omega$
        همانند $\omega$ داریم 
        $X(\omega) = x$،
        که $x$
        یک عدد حقیقی است.
<img src="figs/rv.png" style="float:center; width:50%"/>
        به مثال‌های زیر توجه کنید:
        <ul>
        <li>
        انداختن یک تاس<br/>
        فرض کنید یک تاس را پرتاب کرده‌ایم. نتیجهٔ پرتاب تاس، رو آمدن یکی از وجه‌های تاس است (پس 
        $\Omega$
        مجموعه ای شامل 6 وجه تاس است)
        که روی هر وجه تاس هم تعدادی دایره کشیده شده است. حال یک متغیر تصادفی، می‌تواند این باشد که به هر وجه، تعداد دایره‌هایش را تخصیص دهد. نتیجه این تعریف این است که عدد‌های نسبت داده شده به وجه‌ها، یکی از اعداد 1 تا 6 می‌شود.
 </li>
        </ul>
            <img src="figs/dice4_2.png" style="width:50%;" />   
        احتمالا قبل تر نیز این مدل نگاشت کردن به عدد های 1 تا 6 را دیده بودید. اما متغیر های تصادفی را می‌توان به گونه های دیگری که شاید خیلی واضح نباشند نیز تعریف کرد. مثال ساده‌ی دیگری که می‌توانیم تعریف کنیم زوجیت تعداد دایره‌های روی وجه هاست.
        <ul>
         <li>
        انداختن سکه
        <br/>
        در این حالت، متغیر تصادفی واضحی همچون مثال قبل وجود ندارد. می توان یک متغیر تصادفی را این گونه تعریف کرد.
        <br/>
        $\Omega = {"Head", "Tail"}$
        <br/>
        $X("Head") = 0$
        <br/>
        $X("Tail") = 1$
        <br/>
        </li>
        <li>
             قد دانش آموز
        <br/>
        فرض کنید که یک کلاس داریم و به صورت تصادفی یک شخص را انتخاب می‌کنیم. در نتیجه 
        $\Omega$
        شامل تمام دانش آموزان است. حال متغیر تصادفی می‌تواند این باشد که به هر دانش آموز، قدش را نسبت دهد.
        <br/>
        </li>
    <li>
        در یک آزمایش شامل پرتاب ۵ سکه، تعداد «شیر»ها در این دنباله یک متغیر تصادفی است. البته توجه کنید که خود دنباله یک متغیر تصادفی نیست چون یک مقدار عددی مشخص ندارد.
    </li>
    <li>
        در یک آزمایش ۲ بار یک تاس را پرتاب می‌کنیم. هر یک از موارد زیر یک متغیر تصادفی محسوب می‌شوند:
        <ol>
            <li>
                مجموع عدد حاصل از دو پرتاب
            </li>
            <li>
                تعداد ۶ها در دو پرتاب
            </li>
        </ol>
    </li>
    <li>
        در یک آزمایش شامل انتقال یک پیام، زمان لازم برای انتقال پیام، تعداد نشانه‌های اشتباه دریافت شده و تاخیر دریافت پیام همه متغیرهای تصادفی به حساب می‌آیند.
    </li>
</ul>
حال شاید این سوال ایجاد شود که چرا یک تابع را
"متغیر تصادفی"
می‌نامیم؟
        <br/>
        دلیل آن این است که
چون به خروجی یک آزمایش
تصادفی
        مربوط است،
مقدار آن
تصادفی
است و چون با آن‌ها مانند متغیرهای ریاضی
رفتار می‌کنیم و آن‌ها را با یکدیگر جمع و ضرب می‌کنیم، به توان می‌رسانیم و عملیات ریاضی انجام می‌دهیم، آن‌ها را متغیر نیز می‌نامیم.
        در ادامه‌ی این درس با این عملیات ها نیز آشنا خواهیم شد.
<br />
	</font>
</div>

<div id="sec_def_rv" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
متغیر های تصادفی گسسته و پیوسته
        </font>
		<p></p>
		<hr>
        متغیرهای تصادفی را به صورت کلی به دو نوع پیوسته و گسسته می‌توان تقسیم کرد:
<br/>
        متغیر تصادفی گسسته خروجی‌های مختلف یک آزمایش تصادفی را به زیر مجموعه‌ای شمارا (متناهی یا نامتناهی) از اعداد حقیقی نگاشت می‌کند.
        <br\>
          مثال:
            <ul>
                <li>
                خروجی یک بار پرتاب تاس اعداد طبیعی ۱ تا ۶ است.
                </li>
                <li>
                    تعداد دفعاتی که بایستی یک سکه انداخته شود تا برای اولین بار شیر بیاید، مجموعهٔ همه اعداد طبیعی است که شمارا است.
                </li>
            </ul>
                <br/>
متغیرهای تصادفی پیوسته هم متغیرهایی هستند که خروجی یک آزمایش تصادفی را به زیرمجموعه‌ی ناشمارا از اعداد حقیقی نگاشت می‌کنند. 
           منطقا در این  صورت، مجموعه $\Omega$
                هم ناشمارا است.
         <br/>
معمولاً در این درس مجموعهٔ ناشمارای مهمی که با آن سر و کار داریم، مجموعهٔ اعداد حقیقی است.یعنی به طور ساده می‌توان جملهٔ بالا را این‌گونه تعبیر کرد که اگر آزمایش تصادفی ما خروجی‌ای حقیقی و پیوسته داشته باشد، متغیر تصادفی ما پیوسته است.
         <br/>
          مثال :
            <ul>
                <li>
               به طور مثال همه افراد زنده روی کره زمین در یک لحظه خاص 
را در نظر بگیرید. فردی را به تصادف انتخاب می‌کنیم. اگر دادهٔ مورد علاقه ما سن افراد باشد، متغیر تصادفی‌ای که تعریف می‌کنیم به عنوان ورودی یک انسان می‌گیرد و به عنوان خروجی عدد سن فرد را فرضاً به سال تحویل می‌دهد
(مثلاً فردی می‌تواند
$12.3456$
سال سن داشته باشد).
        اگر فرض کنیم که انسان حداکثر ۱۵۰ سال می‌تواند عمر کند،
برد متغیر تصادفی، اعداد حقیقی بازهٔ
$[0, 150]$
است.
این مجموعه ناشماراست، بنابراین متغیر تصادفی ما پیوسته است.
(به بیان ساده اگر خروجی آزمایش از جنس گسسته مثل 
عدد تاس و شیر یا خط بودن سکه بود، متغیر تصادفی گسسته است. اگر از جنس پیوسته مثل طول و زمان بود، متغیر پیوسته است.)
                </li>
            </ul>
</div>

محاسبهٔ احتمال با استفاده از متغیر تصادفی

		<p></p>
		<hr>
        آزمایش ۳ بار پرتاب یک سکهٔ سالم را در نظر بگیرید. فضای حالات ممکن بدین گونه است:
        <br/>
        $$\Omega=\{HHH, HHT, \dots, TTH, TTT\}$$
        حال فرض کنید متغیر تصادفی
        $X$
        را، تعداد شیرهای واقعهٔ
        $\omega$ تعریف کنیم.
        <br/>
        مقدار این متغیر تصادفی به این صورت خواهد بود:
        </font>

<div id="sec_probability2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
حال در این مثال احتمال این که $X(\omega)\le1$
باشد چیست؟
        <br/>
        احتمال روی همهٔ واقعه‌ها تعریف می‌شود. احتمال $X(\omega)\le1$ یعنی احتمال مجموعهٔ آن $\omega$هایی که برای آن‌ها $X(\omega)\le1$ باشد. یعنی:
$$P\{X\le1\}=P\{\omega:X(\omega)\le1\}=P\{HTT, THT, TTH, TTT\}=\frac{4}{8}$$
دقت کنید که معمولا به طور خلاصه می‌نویسیم: $\{X\le x\}$، ولی این مجموعه‌ای از اعداد نیست، بلکه یعنی: $\{\omega:X(\omega)\le x\}$.
<br/>
        به شکل مشابه برای $X(\omega)=1$ یعنی احتمال مجموعهٔ آن $\omega$هایی که برای آن‌ها $X(\omega)=1$ باشد. یعنی:
$$P\{X=1\}=P\{\omega:X(\omega)=1\}=P\{HTT, THT, TTH\}=\frac{3}{8}$$
حال مثلا اگر احتمال $\{X\le x\}$ را برای هر $x$ بدانیم، احتمال همهٔ واقعه‌های مورد نظر را خواهیم داشت و دیگر لزومی به دانستن فضای احتمال $\Omega$ نیست.(چرا؟)
	</font>
</div>

<p></p>

<div id="sec_intro_pmfcdf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در بخش قبل با تعریف متغیر تصادفی و محاسبهٔ احتمال‌های مرتبط با آن آشنا شدیم و از عباراتی مانند
$P(X(\omega) = x)$
یا
$P(X(\omega) \leq x)$
استفاده کردیم.
        <br />
به عنوان مثال آزمایش تصادفی پرتاب دو تاس متقارن و سالم را در نظر بگیرید. متغیر تصادفی
$X$
را
این‌گونه تعریف می‌کنیم که
به ازای هر
$\omega=(d_1, d_2)$
که
$d_1$
و
$d_2$
عدد خروجی پرتاب تاس اول و دوم هستند،
$X(\omega) = max(d_1, d_2)$.
        برای مثال اگر خروجی یک بار پرتاب همزمان دو تاس اعداد ۴ و ۵ باشند،
$X(\omega)=X((4,5)) = 5$.
<br />
حال فرض کنید که علاقمند به دانستن
این هستیم که
اعدادی که در برد
$X$
قرار دارند
(اعداد طبیعی ۱ تا ۶)
هر کدام چقدر احتمال وقوع دارند.
به عنوان مثال
$P(X(\omega) = 2)$
را می‌خواهیم به دست آوریم.
برای این کار همان‌طور که در مثال جزوه قبلی دیدیم بایستی ابتدا همه
$\omega \in \Omega $
ها را در نظر بگیریم که
$X$
آن‌ها را به ۲ می‌نگارد.
یعنی:
$$\{(1,2), (2,1), (2,2)\}$$
سپس احتمال مد نظر، برابر جمع احتمال تک تک این رویدادهای مجزاست. در نتیجه:
$$
P(X(\omega) = 2) = \frac{1}{36} + \frac{1}{36} + \frac{1}{36} = \frac{3}{36}
$$
حال از اینجا به بعد به جای
$P(X(\omega)=x)$
یا
$P(X=x)$،
به طور خلاصه
از نماد
$p(x)$
استفاده می‌کنیم.
در مثال بالا
مقدار
$p(2)$
برابر
با
$\frac{3}{36}$
بود.
به این تابع،
تابع جرم احتمال می‌گوییم.
	</font>
</div>

<div id="sec_def_pmf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
 تعریف تابع جرمی احتمال (PMF)
        </font>
		<p></p>
		<hr>
        فرض کنید
        $X$
یک متغیر تصادفی و
$x$
یکی از مقادیر ممکن آن باشد.تابع جرم احتمال
$X$
(Probability Mass Function)
        یا به اختصار
        PMF
که با
$p_X(a)$
نمایش داده می‌شود برابر با احتمال وقوع رویداد
$\{X=a\}$
است. یعنی:
        $$p_X(a)=P(\{X=a\})$$
مانند مثال بالا، اگر تنها با یک متغیر تصادفی سر و کار داشته باشیم، می‌توانیم اندیس $X$ را حذف کرده و $p_X(a)$ را به صورت $p(a)$ بنویسیم.
        <br />
توجه کنید که:
<ul>
<li>
    $\forall a \in \mathbb{R}: 0 \leq p(a) \leq 1$
</li>
<li>
$a$
هر عدد حقیقی‌ای می‌تواند باشد. در صورتی که
$a$
جز خروجی‌های ممکن
$X$
نباشد،
$p(a) = 0$
است.
</li>
</ul>
        <br />
در مثال قبلی، مقدار
$p(2)$
را محاسبه کردیم.
مقدار تابع جرم احتمال برای سایر مقادیر ممکن
$X$
در جدول زیر آمده است:
(تابع جرم احتمال به اختصار
pmf
نیز گفته می‌شود.
)

<tr>
    <th>مقدار</th>
    <th>x</th>
    <th>1</th>

خواص تابع توزیع انباشته
        </font>
		<p></p>
		<hr>
     در ادامه به تعدادی از خواص تابع توزیع انباشته اشاره می‌شود:
<ol>
    <li>
$$
\lim_{x\to-\infty}F(x)=0
$$
زیرا:
$$
F(-\infty)=P(\{X\leq-\infty\})=0
$$
    </li>
    <br />
    <li>
        $$
        \lim_{x\to\infty}F(x)=1
        $$
        زیرا:
        $$F(+\infty)=P(\{X\leq+\infty\})=1$$
    </li>
    <br />
    <li>
        $$x_1\leq x_2\Rightarrow F(x_1)\le F(x_2)$$
        این ویژگی به این معناست که $F(x)$ تابعی غیرنزولی است. زیرا:
        $$x_1\leq x_2\Rightarrow\{\omega:X(\omega)\leq x_1\}\subseteq\{\omega:X(\omega)\leq x_2\}\Rightarrow P(\{X\le x_1\})\leq P(\{X\leq x_2\})\Rightarrow F(x_1)\leq F(x_2)$$
به عبارت دیگر تابع توزیع تجمعی به این دلیل که برابر جمع مقادیر مثبت است، هیچ‌گاه کاهش نمی‌یابد.
<br/>
        همچنین از این جا نتیجه می‌شود که: $0\leq F(x)\leq1$، چون دیدیم که در $-\infty$ و $+\infty$ مقدار تابع $F$ به ترتیب صفر و یک است.
    </li>
    <br />
    <li>
    $$P(\{X>x\})=1-F(x)$$
    زیرا:
    $$\{\omega:X(\omega)>x\}=\{\omega:X(\omega)\le x\}^c\Rightarrow P(\{X>x\})=1-P(\{X\le x\})=1-F(x)$$
    </li>
    <br />
    <li>
    $$P(\{x_1<x\le x_2\})=F(x_2)-F(x_1)$$
زیرا:
<br/>
دو مجموعه
$A = \{X\le x_1\}$
و
$B = \{x_1<X\le x_2\}$
را در نظر بگیرید.
توجه کنید که
$A \cup B = \{x_1<x\le x_2\}$
و
$A \cap B = \emptyset$.
<br/>
    $$\{X\le x_2\}=A\cup B \Rightarrow P(\{X\le x_2\})=P(\{X\le x_1\})+P(\{x_1<X\le x_2\})\Rightarrow P(\{x_1<X\le x_2\})=F(x_2)-F(x_1)$$
    </li>
</ol>

            </li>

                چقدر است؟

            </li>

        فرض کنید آزمایشی داریم که نتیجه‌ش به شکل یک عدد است. 
        $n$
        نفر نتیجه‌ی آزمایش را پیش‌بینی کرده‌اند. نتیجه‌ی آزمایش به احتمال یکسان، یکی از اعدادی است که به عنوان پیش‌بینی مشخص شده است.(در واقع نتیجه‌ی آزمایش به شکل یک متغیر تصادفی به نام
        $X$
        است که ما احتمال این که این متغیر تصادفی هر عددش مشخصی را اختیار کند می‌دانیم).
        حال فرض کنید می‌خواهیم نتیجه‌ی آزمایش را حدس بزنیم و یک عدد مثل
        $x$
        اعلام کنیم. تابعی را به عنوان نشان‌دهنده‌ی خطا به این شکل تعریف می کنیم
        $R = (X - x) ^ {2}$
        که 
        $X$
        نتیجه‌ی واقعی آزمایش است. از آن جایی که 
        $X$
        یک متغیر تصادفی است و 
        $R$
        یک تابع از 
        $X$
        است، می‌فهمیم 
        $R$
        نیز یک متغیر تصادفی است و امیدریاضی دارد. حال بر اساس دیتا هایی که داریم، 
        $x$
        را چه قرار دهیم تا امیدریاضی خطا کمینه شود؟
	</font>
</div>
