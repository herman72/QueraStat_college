فرض کنید که 1000 گاو صندوق داریم که هر کدام یک رمز 3 رقمی دارند. احتمال می‌دهیم که رمز حداقل یکی از گاو صندوق ها 007 باشد. برای همین از گاو صندوق شماره ۱ شروع می‌کنیم و به ترتیب، به گاو صندوق ها رمز 007 را وارد می‌کنیم و تا جایی این کار را انجام می‌دهیم که یک گاو صندوق باز شود. دقت کنید که ممکن است هیچ گاو صندوقی باز نشود و در آن صورت هم کار تمام است. حال متغیر تصادفی X  را شماره اولین گاوصندوق باز شده می‌گیریم. حال به پرسش های زیر پاسخ دهید. 
        در صورتی هم که هیچ کدام از گاوصندوق ها باز نشود، متغیر تصادفی مقدار ۱۰۰۱ را اختیار می‌کند.
        <img src="figs/safe.JPG" width="500">
        $$
        \begin{aligned}
        \end{aligned}
        $$
        <ul>
            <li>
                $Pr(X > 1)$ را بیابید.
                <br>
                متغیر تصادفی $X$ از توزیع هندسی با پارامتر $p=\frac{1}{1000}$ پیروی می‌کند. بنابراین احتمال این که شماره صندوقی که با این رمز باز می‌شود بیش از یک باشد برابر با $P(X>1) = \frac{999}{1000}$ می‌باشد.
            </li>
            <li>
فرض کنید تا کنون 500 گاوصندوق را امتحان کرده‌ایم و هیچ کدام باز نشده است. احتمال این که مجبور باشیم بیشتر از 501 گاو صندوق را امتحان کنیم چقدر است ؟    
                <br>
       این به معنی این است که شماره گاوصندوق بازشده بیش از 501 باشد در حالی که می‌دانیم این عدد از 500 بیشتر است. بنابراین:
                <br>
                <div style="text-align: left">$P(X > 501 | X > 500) = \frac{P(X > 501)}{P(X > 500)} = \frac{(\frac{999}{1000})^{501}}{(\frac{999}{1000})^{500}} = \frac{999}{1000}$</div>
             <br>
            </li>
        </ul>
نتیجه دو پرسش بالا این بود که اگر تا کنون هیچ گاو صندوقی باز نکرده باشیم و مجبور باشیم بیشتر از 1 گاو صندوق باز کنیم، همان قدر احتمال دارد که تا کنون 500 گاو صندوق را تست کرده باشیم و مجبور باشیم بیشتر از 1 گاو صندوق "دیگر" را تست کنیم. در واقع انگار گاو صندوق های قبلی تاثیری بر این که گاو صندوق های بعدی باز می شوند یا خیر ندارند. به این ویژگی از این توزیع، خاصیت بی حافظگی می گویند.

        <div style="text-align: left"> $Pr[X > s+t | X > t] = Pr[ X > s], s,t > 0 $ </div>
        $$
        \begin{aligned}
        \end{aligned}
        $$
در صورتی که $X$ را به صورت طول عمر یک دستگاه در نظر بگیریم معادله بالا بیانگر آن است که احتمال این که این دستگاه حداقل به اندازهٔ $s + t$ ساعت عمر کند با توجه به این که $t$ ساعت عمر کرده برابر با احتمال اولیه‌ی این است که حداقل $s$ ساعت عمر کند. به بیان دیگر اگر دستگاه در زمان $t$ هنوز سالم باشد توزیع طول عمر باقیمانده‌ی آن همان توزیع طول عمر اولیه‌ی دستگاه میباشد. (دستگاه فراموش میکند که به مدت $t$ استفاده شده است). درواقع بی حافظه بودن به این معناست که رویداد های قبلی تاثیری بر احتمال رویداد های آینده ندارد.

                مثال بی‌حافظگی
        </font>
		<p></p>
		<hr>
        <ul>
            <li>شخصی می‌خواهد به مسافرت برود و به یک شرکت اجاره ماشین می‌رود تا ماشینی بگیرد. او می‌داند که طول عمر باتری موجود در ماشین‌ها از توزیع نمایی پیروی می‌کنند. بر روی باتری هر ماشین میانگین طول عمر آن و بر روی کیلومترشمار آن مسافت طی شده توسط ماشین نوشته شده است. حال با توجه به اطلاعات در دسترس، شخص کدام ماشین را انتخاب کند که به احتمال بیشتری بدون تمام شدن باتری به مقصد برسد؟ 
                <br>
                <br>
                توزیع نمایی یک توزیع بی‌حافظه است. زیرا:
                <br>
                <br>
                <div style="text-align: left">
                    $Pr[X > s+t | X > t] = \frac{Pr[X > s+t, X > t]}{Pr[x>t]} = \frac{1-F(s+t)}{1-F(t)} = \frac{e^{s+t}}{e^{t}} = e^s = Pr[X > s]$
                </div>
                 $$
                \begin{aligned}
                \end{aligned}
                $$
                                بنابراین مسافت طی شده توسط ماشین‌ها اهمیتی ندارد و بر احتمال تمام شدن باتری در طی مسافت $s$ تاثیرگذار نیست. در نتیجه بهترین انتخاب، باتری با میانگین طول عمر بیشتر است هر چند که بیشتر از همه کار کرده باشد.                
                <br><br>
                توجه:در دنیای واقعی عمر باتری ماشین از توزیع نمایی پیروی نکرده و با گذر زمان احتمال بقای آن کمتر میشود اما از مثال‌های بی‌حافظگی میتوان به نیمه عمر مواد رادیواکتیو ،زمان کشف بلاک جدید بیت‌کوین و یا زمان رسیدن مشتری بعدی در نظریه‌ی صف اشاره کرد.
                <br><br>
            </li>
            <li>
               شخصی به اداره ای می‌رود که دو منشی دارد .میزانی که طول می‌کشد تا هر منشی، کار یک ارباب رجوع را انجام دهد از یک توزیع نمایی با پارامتر 
                $\lambda$
                پیروی می‌کند. زمانی که فرد مورد نظر به آنجا می‌رسد هر دو منشی مشغول انجام دادن کار یک ارباب رجوع هستند و به محض این که کار هر یک تمام شود نوبت به کار شخص می‌رسد. احتمال این که این شخص آخرین نفری باشد که کارش تمام می‌شود چقدر است؟
                <br><br>
                از آن جا که توزیع نمایی بی حافظه است زمانی که نفر اول اعترافش تمام شود مانند این است که شخص مورد نظر و فرد دیگر با توزیع یکسانی همزمان به سراغ منشی ها رفته اند. پس احتمال این که کار شخص دیرتر تمام شود برابر با $\frac{1}{2}$ می‌باشد. البته ما فرض کرده‌ایم که حتما یکی زود تر کارش به پایان می‌رسد. چرا این فرض احتمال را تغییر نمی‌دهد؟     
            </li>
        </ul>

            تمرین تئوری
        </font>
        <p></p>
        <hr>
        <li>
            فرض کنید که 
            $X$
            یک متغیر تصادفی بی‌حافظه است. 
            $F_X$
            را تابع 
            CDF
            این متغیر تصادفی و 
            $P_X$
            را تابع 
            PMF
            این متغیر تصادفی می‌نامیم. 
            <br>
-	
            $P(X >= j+k)$ را برحسب 
            $F_X(j)$
            ، $F_X(k)$
            ، $P_X(j)$
            و $P_X(k)$
            بنویسید.
            <br>
-	یک توزیع بی‌حافظه نام ببرید. آیا مقداری که در بالا بدست آورده‌اید، در توزیع مذکور صدق می‌کند؟
        </li>
        <li>
            فرد می‌خواهد ماشیشنش را بفروشد. برای این کار، ماشین را به اولین کسی که قیمت بیشتر از ۱۵۰۰۰ دلار پیشنهاد دهد، می‌فروشد. پیشنهاد‌ها هم از توزیع نمایی با میانگین ۱۰۰۰۰ دلار پیروی می‌کنند. 
            $X$
            را تعداد پیشنهاد هایی که فرد رد خواهد کرد می‌نامیم. 
            $E[X]$ 
            چند است؟ 
            $Y$
            را نیز پولی که فرد دریافت می‌کند می‌نامیم. 
            $E[Y]$
            چند است؟
        </li>
        <li>
            جو منتظر آمدن قسمت جدید جدید تام و جری است. او اطلاعاتی در مورد زمان پخش این کارتون دارد. به این صورت که اگر مقیاس را سال در نظر بگیریم و زمان آمدن قسمت جدید را 
            $T$
            تعریف کنیم، 
            $T \sim Exp(\frac{1}{5})$
            . برای آگاه شدن از این که قسمت جدید آمده است یا خیر، جو در آخر هر روز خبرهای مربوط به دنیای سینما را بررسی می‌کند. برای مثال، اگر کارتون در ظهر روز دوم بیاید، جو در آخر روز دوم متوجه می‌شود. روزی که جو از آمدن کارتون آگاه می‌شود را 
            $X$
            می‌نامیم. تابع 
            PMF
            را برای 
            $X$
            بیابید. آیا 
            $X$
            از توزیع معروفی پیروی می‌کند؟
        </li>
        <li>
            توزیع هندسی (گسسته) و توزیع نمایی (پیوسته) تنها توزیع‌های بی‌حافظه می‌باشند. ابتدا بی حافظگی آنها را بررسی کنید و سپس سعی کنید یکتایی آن‌ها را اثبات کنید.        </li>
        </font>
</div>
