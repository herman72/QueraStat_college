<div id="normal_attributes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            توزیع نرمال
        </font>
		<p></p>
        <hr>
        توزیع نرمال یا گوسی، مهم ترین توزیع احتمالاتی در آمار می‌باشد. زیرا در بررسی های موارد پدیده های مختلف، می‌یابیم که این پدیده ها از این توزیع پیروی می‌کنند. برای مثال، قد افراد، فشار خون، خطا های اندازه گیری، و میزان IQ افراد از مواردی هستند که از این توزیع پیروی می‌کنند.
        <br>
        از آنجا که توزیع نرمال توزیعی پیوسته است،‌ تابع PMF آن در تمامی نقاط صفر است. بنابراین برای محاسبات،‌ از تابع چگالی یا PDF استفاده می کنیم. البته می توان از تابع احتمال تجمعی یا CDF نیز استفاده کرد اما کار با PMF راحت‌تر و مرسوم‌تر است.
        توزیع نرمال دو پارامتر دارد که با 
        $\mu$
        و
        $\sigma^{2}$
        نشان می دهیم.
        حال اگر متغیر تصادفی 
        $X$
        از توزیع نرمال با پارامترهای مشخض پیروی کند (که به شکل
        $X \sim N(\mu, \sigma^{2})$
        نشان می‌دهیم) داریم :        <br>
        $$ f(x; \mu,\sigma ^{2}) = \frac{1}{\sqrt{2\pi\sigma^{2}}} e ^{\frac{-(x-\mu) ^ {2}}{2\sigma ^{2}}}$$
        در قسمت زیر می‌توانید نمودار 
        $f(x) - x$
        را برای این توزیع با پارامترهای مختلف مشاهده کنید.
        <img src="figs/Normal_Distribution_PDF.png" />
	</font>
</div>

<br/>
<div id="sec_refs" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
                امید ریاضی و واریانس توزیع نرمال
        </font>
		<hr>    
        محاسبهٔ امیدریاضی و واریانس توزیع نرمال خیلی ساده نیست و در این جا تنها به ذکر این موارد می‌پردازیم.
        اگر متغیر تصادفی
        $X$
        از توزیع نرمال با پارامتر های 
        $\mu$
        و 
        $\sigma^{2}$
        پیروی کند، داریم :
        <ul>
            <li>
                امیدریاضی: 
                $$ E[X] = \mu$$
            </li>
            <li>
                واریانس:
                $$ var[X] = \sigma^{2}$$
            </li>
        </ul>
        در واقع می‌بینیم که پارامترهای توزیع نرمال، همان امیدریاضی و واریانس توزیع هستند. بنابراین برای توصیف یک توزیع نرمال، کافی است تا امیدریاضی و واریانس توزیع را مشخص کنیم و پس از آن، می‌توانیم در هر نقطه مثل 
        $x$
        ، مقدار تابع چگالی را بدست آوریم.
        <br>
        هم چنین در مورد توزیع نرمال، مورد زیر را می دانیم:
                اگر $X \sim N(\mu,\sigma ^{2})$ و a و b دو عدد حقیقی باشند؛ آنگاه:
                $$ aX + b \sim N(a\mu + b, (a\sigma)^{2}) $$
                زیرا با افزودن این اعداد، توزیع نهایی هم چنان نرمال می‌ماند(چرا؟) و می‌دانیم که با ضرب عدد در یک توزیع یا جمع یک عدد 
                با توزیع، چه تفاوتی در امیدریاضی و واریانس متغیر بدست آمده ایجاد می‌شود.
    </font>
</div>


<div id="sec_refs" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            بدست آوردن تقریبی تابع احتمال تجمعی
        </font>
		<hr>    
        توزیع نرمال با امید ریاضی 0 و واریانس 1 را توزیع نرمال استاندارد می‌گویند و آن را با Z نشان می‌دهیم.
        تابع توزیع تجمعی(CDF) در توزیع نرمال استاندارد با $\phi$ نشان داده می‌شود.
        $$ F(x) = \phi(x) $$
        تابع توزیع تجمعی فرمول تمیزی ندارد و از جدول $\phi$ برای آن استفاده می‌کنیم.
        <br>
        <br>
        در حالت کلی برای متغیر تصادفی X با توزیع نرمال $N(\mu,\sigma^{2})$ طبق فرمول ترکیب خطی داریم:
        $$ Z = \frac{X - \mu}{\sigma} $$
        $$ F(x) = P(X \le x) = P(\frac{X - \mu}{\sigma} \le \frac{x - \mu}{\sigma}) = \phi(\frac{x - \mu}{\sigma}) $$
        <br>
        <img src="figs/phi_table.png" />
        برای اعداد منفی هم با توجه به تقارن در توزیع نرمال استاندارد نسبت به میانگین توزیع داریم:
        $$ \phi(-x) = 1 - \phi(x) $$
        <br>
        </font>
</div>
 <br>

<div id="normal_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            مثال توزیع نرمال
        </font>
		<p></p>
        <hr>
        اگر X یک متغیر تصادفی نرمال با پارامتر های $\mu = 3$ و $\sigma^{2} = 9$ باشد احتمال $P(2 < X < 5)$ چقدر است؟
        <br>
        <br>
        $$ P(2 < X < 5) = P(\frac{2 - 3}{3} < \frac{X - 3}{3} < \frac{5 - 3}{3}) $$
        $$ = P(-\frac{1}{3} < Z < \frac{2}{3}) $$
        $$ = \phi(\frac{2}{3}) - \phi(-\frac{1}{3}) $$
        $$ = \phi(\frac{2}{3}) - [1 - \phi(\frac{1}{3})] $$
        $$ \approx 0.7486 - [1 - 0.6293] = 0.3779 $$
<br>
	</font>
</div>

            منشی نمونه
        </font>
		<p></p>
        <hr>
        اردشیر به تازگی منشی یک شرکت  شده است. به او گفته شده است که برای حفظ پرستیژ شرکت تازه تاسیس، تعداد افرادی که در هر ساعت به کار آن‌ها رسیدگی می‌شود باید از توزیع یکنواخت بین m تا n نفر
         پیروی کند. اما متاسفانه کامپیوتر او این محاسبات را انجام نمی‌دهد و حتی نمی‌تواند عدد رندوم تولید کند. 
اما اردشیر که به‌ درس آمار و احتمال بسیار مسلط بود، می‌دانست فاصله‌ی زمانی افرادی که به او مراجعه می‌کنند، از توزیع نمایی پیروی می‌کند و پارامتر آن را پس از مدتی به‌دست آورد. حال می‌داند که می‌تواند از این نمونه‌ها، نمونه‌های توزیع خواسته شده را به‌دست آورد. به نظر شما او چطور این کار را انجام می‌دهد؟

<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            مقدمه
        </font>
		<p></p>
        <hr>
        در قسمت‌های قبلی با متغیر تصادفی آشنا شدیم. می‌دانیم متغیر تصادفی یک نگاشت از فضای نمونه به اعداد حقیقی است. 
        $$ X: \Omega \rightarrow \mathbb{R} $$
        <img src="figs/RV.png" style="float:center; width:40%"/>
        <br>
        حال می‌خواهیم با تغییر (transformation)  های متغیرهای تصادفی آشنا شویم. که ساده‌ترین آن‌ها، تابعی از یک متغیر تصادفی است. 
        <br>
        Y را به صورت تابعی از متغیر تصادفی پیوسته‌ی X به شکل زیر تعریف می‌کنیم.
        <br>
        $$ \textbf{Y} = g(\textbf{X}) $$
        $$ g: \mathbb{R} \rightarrow \mathbb{R} $$
        <br>
        که در واقع خلاصه‌شده‌ی تعریف زیر است:
        $$  \textbf{Y} (\omega) = g(\textbf{X}(\omega)) , \ \omega \in \Omega $$
        <br>
        در این‌جا، Y نیز یک متغیر تصادفی پیوسته خواهد بود و توزیع خاص خود را خواهد داشت. 
        <br>
        <img src="figs/y_omega.png" style="float:center; width:50%"/>
        <br> <br>
        در این درس دو روش برای به‌دست آوردن تابع چگالی متغیر تصادفی Y معرفی می‌کنیم. اول، روی g های یک به یک متمرکز شده و پس از آن بقیه توابع را بررسی خواهیم کرد. در آخر نیز یاد می‌گیریم چطور با استفاده از معکوس توابع توزیع تجمعی، می‌توان از هر توزیعی که در دسترس داریم، توزیع دیگری را بسازیم.
	</font>
</div>

<div id="sec_CDF" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
        توزیع تجمعی تابعی از یک متغیر تصادفی(روش نخست)
        </font>
        <hr>
        برای به دست آوردن توزیع $ Y = g(X) $ می‌توان به‌صورت مستقیم و با استفاده از تعریف تابع توزیع تجمعی، این مقدار را به‌دست آورد. یعنی به طور کلی به‌صورت زیر عمل کرد:
        <br> 
$$ F_Y(y) = P(Y \le y) = P(g(X) \le y) = \sum P(x_i \le X \le x_j) = \sum \pm \ F_X(x_i) $$
        <br>
در حالت خاص اگر تابع g یک به یک و صعودی باشد خواهیم داشت:
        <br>
$$ F_Y(y) = P(Y \le y) = P(g(X) \le y) = P(X \le g ^{-1} (y)) = F_X(g ^{-1} (y)) $$
        <br>
        پس برای تابع توزیع چگالی Y در این حالت خواهیم داشت:
        <br>
        $$ f_Y(y) = \frac{d F_Y(y)}{dy} =\frac {d F_X(g ^{-1} (y))}{dy} = \frac{f_x(g ^{-1} (y))}{g'(g ^{-1} (y))} $$
        <br>
        محاسبات مربوط به حالت کلی در بخش بعد آمده است. 
	</font>
</div>



<p></p>
<br/>
<div id="sec_CDF" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
        مثال توزیع تجمعی تابعی از یک متغیر تصادفی(روش نخست)
        </font>
        <hr>
        <ul>
            <li>
                فرض کنید X یک متغیر تصادفی پیوسته با توزیع معلوم است. توزیع Y را به‌دست آورید. 
                $$ Y = aX + b $$
                <br> 
                با استفاده از روش معرفی شده در بالا به‌صورت زیر عمل می‌کنیم.
                <br>
                $$ \begin{align*}
                F_Y(y) &= P(Y \le y) = P(aX+b \le y)  \\
                &\overset{a > 0}{=} P(Y \le \frac{y - b}{a}) = F_X(\frac{y - b}{a})  \\
                &\overset{a < 0}{=} P(Y \ge \frac{y - b}{a}) = 1 - F_X(\frac{y - b}{a})
                \end{align*} $$
                <br>
            </li>
            <li>
                فرض کنید X یک متغیر تصادفی پیوسته با توزیع معلوم است. توزیع Y را به‌دست آورید. 
                $$ Y = X^2  $$
                <br> 
                همانند مثال قبل داریم:
                <br>
                $$ F_Y(y) = P(Y \le y) \overset{y > 0}{=} 
            P(-\sqrt{y} \le X \le \sqrt{y}) = F_X(\sqrt{y}) - F_X(- \sqrt{y}) $$
                <br>
                همچنین می‌توان تابع توزیع چگالی را به‌روش زیر به‌دست آورد:
                <br>
                $$ \frac{d F_X(\sqrt{y})}{dy} = \frac{1}{2\sqrt{y}} f_X(\sqrt{y}) \Rightarrow
                 f_Y(y) = \frac{1}{2\sqrt{y}} f_X(\sqrt{y}) + \frac{1}{2\sqrt{y}} f_X(-\sqrt{y}) , \ (y>0) $$
                <br> 
                <img src="figs/X2.png" style="float:center; width:80%"/>
            </li>
        </ul>
    </font>
</div>

<div id="sec_PDF" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
                بدست آوردن تابع توزیع چگالی به صورت مستقیم(روش دوم)
        </font>
        <hr>
        <font color=#FF7500>
                قضیه:
        </font>
        اگر X یک متغیر تصادفی باشد و متغیر تصادفی Y از رابطهٔ $Y=g(X)$ بدست آمده باشد
و
$x_1, x_2, … $ 
ریشه‌های معادله‌ی $y=g(x)$ باشند. (یعنی:
$ y = g(x_1) = … = g(x_i) = … $)
        <br>
        تابع توزیع چگالی Y در نقطه‌ی y از رابطه‌ی زیر به‌دست می‌آید.
        <br>
        <br>
        $$ f_Y(y) = \frac{f_X(x_1)}{|g'(x_1)|} + ... + \frac{f_X(x_i)}{|g'(x_i)|} + ... = \sum_i \frac{f_X(x_i)}{|g’(x_i)|} $$
        <br>
        (اثبات قضیه را می‌توانید در صفحه‌ی ۱۱۸، بخش ۴-۴ منبع اول پیدا کنید.)
    <br>
    <img src="figs/pdf_theorem.png" style="float:center; width:50%"/>
    <br>
    <br>
    <font color="blue">
            مثال:
    </font>
فرض کنید یک لیزر را که در فاصله‌ی واحد از مبدا قرار دارد، (مطابق شکل) به زمین می‌تابانیم. متغیر تصادفی $X$ بیانگر فاصله افقی نقطه‌ی برخورد این پرتو نسبت به مبدا در آزمایش‌های متوالی است. اگر بدانیم 
$\theta$
که بیانگر زاویه قرار گیری لیزر است از توزیع یکنواخت 
$(- \pi / 2, \pi / 2)$
پیروی می‌کند، بررسی کنید $X$ از چه توزیعی است. 
    <img src="figs/cauchy.png" style="float:center; width:50%"/>
     حل: ابتدا متغیر تصادفی $X$ را به‌صورت تابعی از متغیر تصادفی
     $\theta$
     که توزیع آن را داریم می‌نویسیم:
     $$ X = tan(\theta)$$
   حال با استفاده از روش بالا خواهیم داشت:
   $$ f_X(x) = \frac{f_{\theta} (\theta _0)}{1 + tan^2 \theta_0} , \;\; \theta_0 = tan^{-1} x 
   \Rightarrow f_X(x) = \frac{\frac{1}{\pi}}{1 + x^2} =  f_X(x) = \frac{1}{\pi(1 + x^2)} $$
	</font>
</div>



<p></p>
<br/>
<div id="sec_PDF" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
                مثال بدست آوردن تابع توزیع چگالی به صورت مستقیم(روش دوم)
        </font>
        <hr>
فرض کنید یک لیزر را که در فاصله‌ی واحد از مبدا قرار دارد، (مطابق شکل) به زمین می‌تابانیم. متغیر تصادفی $X$ بیانگر فاصله افقی نقطه‌ی برخورد این پرتو نسبت به مبدا در آزمایش‌های متوالی است. اگر بدانیم 
$\theta$
که بیانگر زاویه قرار گیری لیزر است از توزیع یکنواخت 
$(- \pi / 2, \pi / 2)$
پیروی می‌کند، بررسی کنید $X$ از چه توزیعی است. 
    <img src="figs/cauchy.png" style="float:center; width:50%"/>
     حل: ابتدا متغیر تصادفی $X$ را به‌صورت تابعی از متغیر تصادفی
     $\theta$
     که توزیع آن را داریم می‌نویسیم:
     $$ X = tan(\theta)$$
   حال با استفاده از روش بالا خواهیم داشت:
   $$ f_X(x) = \frac{f_{\theta} (\theta _0)}{1 + tan^2 \theta_0} , \;\; \theta_0 = tan^{-1} x 
   \Rightarrow f_X(x) = \frac{\frac{1}{\pi}}{1 + x^2} =  f_X(x) = \frac{1}{\pi(1 + x^2)} $$
	</font>
</div>

<div id="sec_ITM" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
                روش تبدیل معکوس (یا Inverse Transform Sampling)
        </font>
        <hr>
        فرض کنیم منبعی وجود داشته باشد که بتوانیم با آن نمونه‌هایی از توزیع پیوسته‌ی X با تابع توزیع تجمعی $F_X$  به‌دست آوریم. و می‌خواهیم نمونه‌هایی از توزیع مشخص Y داشته باشیم. یعنی می‌خواهیم تابع g رابه گونه‌ای بدست بیاوریم که بتوانیم با نمونه‌هایی از توزیع X، نمونه‌های توزیع Y را بسازیم. یعنی:
<br>
        $$ Y = g(X) $$
        برای این کار ۳ حالت مختلف را بررسی می‌کنیم.
                <ul>
            <li>
                بدست آوردن توزیع یکنواخت پیوسته از روی توزیعی دلخواه
                <br>
                فرض کنید $F$ تابع توزیع تجمعی متغیر تصادفی X باشد. می‌دانیم 
                $ F: \mathbb{R} \longrightarrow [0, 1] $.
                حال اگر
                $Y = F(X)$
                باشد،
                $ 0 \le Y \le 1 $.
                حال می‌خواهیم برای Y، تابع توزیع تجمعی $F$ را به‌دست آوریم.
                <br><br>
                $$ F_Y(y) = \mathbb{P}(Y \le y) = \mathbb{P} (F(X) \le y) =  \mathbb{P} (X \le F^{-1}(y)) =
                F_X(F^{-1}(y)) = y $$
                <br>
                بنابراین Y توزیع یکنواخت بین $[0, 1]$ است. 
            </li>
                    <br>
            <li>
                بدست آوردن توزیعی دلخواه از روی توزیع یکنواخت پیوسته
                <br>
                اگر فرض کنیم U تابع توزیع یکنواخت بین $[0, 1]$ و h یک تابع اکیداً صعودی با برد $[0, 1]R =  $  باشد. می‌خواهیم ثابت کنیم که اگر داشته باشیم:
                $$ Y = h^{-1}(U) $$ 
                آن‌گاه h تابع توزیع تجمعی Y است. 
                <br><br>
                می‌دانیم
                h صعودی است. بنابراین $h^{-1}$ نیز صعودی است. پس داریم:
            <br>
            $$ F_Y(y) = \mathbb{P}(Y \le y) = \mathbb{P} (h^{-1}(U) \le y) \overset{h^{-1}\text{ is increasing}}{=} \mathbb{P} (U \le h(y)) =  F_U(h(y)) \overset{\text{U is uniform}}{=} h(y) $$
            <br>
            </li>
            <li>
                داشتن توزیع دلخواه 
                $X$
                و بدست آوردن توزیع دلخواه
                $Y$
                از روی آن
                <br>
                در نهایت فرض می‌کنیم توزیع X را داریم. برای تابع دلخواه صعودی $F_Y$ با شرط‌های  
                $ lim_{x \rightarrow\infty} F_Y(x)dx = 1$ و $ lim_{x \rightarrow- \infty} F_Y(x)dx = 0 $
                ، می‌خواهیم تابع g در $Y = g(X)$ را به گونه‌ای بدست آوریم که این تابع، CDF متغیر تصادفی Y باشد. 
                <br><br>
                با استفاده از قسمت ۱ می‌دانیم توزیع U به صورت زیر بدست می‌آید:
                $$ U = F_X(X) $$
                <br>
                همچنین از قسمت ۲ می‌دانیم:
                $$ Y = F_Y^{-1} (U) $$
                <br>
                بنابراین خواهیم داشت:
                $$ Y =  F_Y^{-1} (F_X(X)) $$
                <br> <br>
                می‌توان به‌روش زیر نیز تابع g را به‌دست آورد:
                $$ F_Y(y) = \mathbb{P}(Y \le y) = \mathbb{P} (g(X) \le y)
                = \mathbb{P} (X \le g^{-1} (y)) = F_X(g^{-1} (y))
                 \Rightarrow y = F_Y^{-1}(F_X(g^{-1} (y))) 
                 \Rightarrow g(y) = F_Y^{-1}(F_X(y)) $$
                <br><br>
            </li>
        </ul>
    </font>
</div>

		<font color=#FF7500 size=6>

	</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
m = 5
n = 10
scale = 1
N = 10000
X = np.random.exponential(scale, N)
Y = scale * np.exp(-X)
Y = (n - m) * Y + m
plt.hist(Y, 100, alpha=0.3, color="red")
plt.hist(np.random.uniform(m, n, N), 100, alpha=0.3, color="darkblue")
plt.show()
```

            <li>
                فرض کنید که در یک مسابقه، سه داور نمره به شرکت‌کنندگان نمره می‌دهند. نمره‌ی هر داور یک عدد تصادفی بین ۰ تا ۱۰ است(اعداد حقیقی). امتیاز یک نفر ماکسیمم امتیازی است که از ۳ نفر گرفته است.
                <br>
-	احتمال این که داور اول بیشترین نمره را بدهد چقدر است؟
                <br>
                -	اگر 
                $X$
                امتیاز یک شرکت کننده باشد، 
                $E[X]$
                چند است؟
                (چرا 
                $X$
                متغیر تصادفی است؟)
            </li>
            <li>
                فرض کنید که 
                $Z \sim N(0,1)$
                . حال یک دستگاه داریم که مقدار 
                $Z$
                را به ما نشان می‌دهد. اما این دستگاه تنها اعداد مثبت یا صفر را به ما نشان می‌دهد. درنتیجه اگر مقدار 
                $Z$
                مثبت باشد، خود آن مقدار را نشان می‌دهد و در غیر این‌صورت ۰ نشان می‌دهد. 
                $E[X]$
                و 
                $var[X]$
                را بدست آورید.
            </li>
