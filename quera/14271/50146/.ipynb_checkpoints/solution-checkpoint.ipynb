<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>

مقدمه
        </font>
		<p></p>
		<hr>
در نظریه احتمال سر و کله‌ی مجموعه‌ها پیدا می‌شود. در این قسمت ابتدا می‌خواهیم ردپای نظریه‌ی مجموعه‌ها را در احتمالات دنبال کنیم.
سپس اجزای مختلف فضای احتمال را معرفی می‌کنیم و مفهوم احتمال برای مجموعه‌های مختلف را مطالعه خواهیم کرد.
<br/>
یکی از راه‌ها برای فکر کردن درباره‌ی امر تصافی بودن این است که سعی کنیم راهی پیدا کنیم تا چگونه چیزی را که نمی‌دانیم ابراز کنیم. وقتی می‌گوییم چیزی تصادفی است می‌خواهیم بگوییم که دانش ما در مورد پیشامد محدود است، پس با قطعیت نمی‌توان گفت چه اتفاقی خواهد افتاد.
<br/><br/>
	</font>
</div>

<p></p>
<br />
<div id="pr_ex" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#0f6d02 size=6>
مثالی از کاربرد احتمالات
        </font>
		<p></p>
		<hr>
        مخابرات در زندگی روزمره ما نقش ویژه‌ای ایفا می‌کند. برای مثال هنگامی که با تلفن همراه خود صحبت می‌کنید، هنگامی که به شبکه اینترنت متصل می‌شوید و یا وقتی که از یک کنترل از راه دور برای کنترل استفاده می‌کنید از مخابرات کمک می گیرید. همه این کارها بر انتقال اطلاعات استوار هستند. برای مثال وقتی شما با تلفن همراه خود صحبت می‌کنید صدای شما به رشته‌ای از بیت‌ها تبدیل می‌شود که به آن‌ها information bits گفته می‌شود.
        <br/>
        مشکلی در ارسال بیت‌ها وجود دارد آن هم این است که نویز این روند را تحت تاثیر قرار می‌دهد و ممکن است برای مثال رشته "...۱۰۱۰" به رشته "...۱۰۱۱" تبدیل شود که در این جا بیت چهارم دچار خطا شده است. این خطا‌ها کیفیت مخابرات را تحت تاثیر قرار می‌دهند و مثلا می‌تواند باعث شود صدای شما کیفیت خود را از دست بدهد.
        <br/>
        نویز یک پدیده تصادفی است که ما قبل از ارسال بیت‌ها نمی‌دانیم کدام بیت ممکن است دچار خطا شود. از تئوری احتمالات در مخابرات به شدت استفاده می‌شود برای مثال می‌توان برای مدل کردن رفتار نویز به احتمالات رجوع کرد و درنهایت حتی می توان سیستمی در نظر گرفت تا بیت‌هایی که دچار خطا می‌شوند را تصحیح کرد.
<br/><br/>
	</font>
</div>

<div id = "prob_ex" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=0f6d02 size=6>
پیدایش احتمال 
        </font><br/>
        <hr>
        ریاضی‌دان‌ها به دنبال ارائه‌ای از اصولی بودند تا بتوانند مسائل احتمالی را توصیف و فرمول بندی کنند. مثل اصول هندسه اقلیدس یا اصول مکانیک کلاسیک نیوتون.
        <br/>
        یک تعریف پیشنهادی برای احتمال پیشامد $A$ که به ذهن می رسد نسبت تعداد دفعاتی است که پیشامد $A$ در $n$ آزمایش رخ می‌دهد.
        <br/>
        برای مثال اگر یک سکه را ۱۰۰۰ دفعه پرتاب کنیم و ۸۰۰ مرتبه سکه شیر بیاید و ۲۰۰ بار خط بیاید، می‌توانیم بگوییم احتمال شیر آمدن در پرتاب سکه برابر ۰.۸ است، یا به عبارتی اگر این سکه را ۱۰ بار دیگر پرتاب کنیم انتظار داریم ۸ مرتبه به شیر برسیم و ۲ بار خط را ببینیم. البته به طور حتم این استدلال دقیق نیست. مثلا ممکن است سکه را ۱۰۰۰۰ مرتبه پرتاب کنیم و ببینیم ۷۵۰۰ بار شیر آمد و ۲۵۰۰ مرتبه خط. پس نیاز داریم استدلال خود را بسته به دقتی که به دنبال آن هستیم تصحیح کنیم.
        <br/>
        به نظر می‌رسد هر چه $n$ را بزرگ‌تر کنیم به نتیجه دقیق‌تری خواهیم رسید. بر این اساس عده‌ای تعریف زیر را برای احتمال در نظر گرفتند:
        <br/>
        $Pr(A) = \lim_{n \to \infty} \frac{n(A)}{n}$
        <br/>
        اما این تعریف چند اشکال داشت: 
        <ul>
        <li>
        این حد ممکن است وجود نداشته باشد.
        </li>
        <li>
        شاید این حد مقدار ثابتی را ندهد، یعنی در آزمایش‌های مختلف جواب‌های گوناگونی بگیریم.
        </li>
        <li>
        این که $n$ را به سمت بی‌نهایت میل دهیم در دنیای واقعی امکان ناپذیر است.
        </li>
        </ul>
        امروزه ریاضی‌دان ها از اصول کولموگروف برای برخورد با احتمالات بهره می‌گیرند که در ادامه آن‌ها را مطرح می‌کنیم.
        <br>
        <br/>
        
</div>

<div id="review" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>

یادآوری نظریه مجموعه ها
        </font>
		<p></p>
        <hr>
        <ul>
        <li>
       مجموعه (Set) : مجموعه دسته‌ای از اشیا می باشد و هر شی در آن دسته عضوی از مجموعه می‌باشد. برای عضو $x$ که در مجموعه $S$ قرار دارد می‌گوییم $x$ عضو $S$ است و با $x ∈ S$ نمایش می‌دهیم.
       </li>
       </ul>
       <ul>
       <li>
        عضویت: اگر شی $x$ در میان اشیائی که در $S$ هستند نباشد آن‌گاه $x$ عضو $S$ نیست و می‌نویسیم $x \notin S$.
        </li>
        </ul>
        <ul>
        <li>
        تعداد اعضای مجموعه می‌تواند محدود، نامحدود و ناشمارا(متناظر با اعداد طبیعی) و یا غیر‌قابل شمارش باشد. ترتیب اعضا در مجموعه نیز اهمیتی ندارد.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه $B$ را زیرمجموعه‌ مجموعه‌ $A$ می‌گوییم اگر و تنها اگر هر عضو از مجموعه $B$ عضوی از مجموعه $A$ باشد می‌نویسیم $B \subset A$.
        در غیر این صورت $B$ زیرمجموعه $A$ نیست و می‌نویسیم: $B \not\subset A$.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه $A$ را مساوی مجموعه $B$ می‌گوییم اگر و تنها اگر $A \subset B$ و $B \subset A$.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه‌ای که شامل همه‌ی عناصر موجود است را مجموعه مرجع $\Omega$ می‌گوییم.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه‌ای که عضوی ندارد را تهی می‌نامیم و با $\varnothing$ نمایش می‌دهیم.
        </li>
        </ul>
        <ul>
        <li>
       اجتماع دو مجموعه $A$ و $B$ مجموعه تمام عناصری است که در $A$ یا در $B$ یا در هر دوی آن‌ها باشند و اجتماع دو مجموعه را با $A \cup B$ نمایش می‌دهیم.
       </li>
       </ul>
       <img src="Venn_A_union_B.png" style="float:center; width:50%"/>
       <ul>
       <li>
           اشتراک دو مجموعه (مجموعه‌ی
           $A \cap B$) مجموعه عناصری است که هم در $A$ و هم در $B$ هستند.
       </li>
       </ul>
       <img src="A_int_B.png" style="float:center; width:50%"/>
       <ul>
       <li>
       دو مجموعه را جدا از هم (Disjoint) می‌گوییم اگر و تنها اگر $A \cap B = \varnothing$ باشد.
       </li>
       </ul>
       <img src="dis.png" style="float:center; width:50%"/>
	</font>

نمودار ون 
        </font><br/>
        نمودار وِن یا نمودار؛ مجموعه تصاویری است که در شاخه‌ای از ریاضیات به نام نظریه مجموعه‌ها به کار می‌رود که در سال ۱۸۸۱ توسط جان ون اختراع شد. این نمودار تمام روابط ممکن ریاضی یا منطقی مابین مجموعه‌ها را نشان می‌دهد و معمولاً از دایره‌های هم‌پوشان تشکیل شده است. نمودار ون به نسبت تعداد دوایر موجود در آن به چند بخش تقسیم می‌شود. برای مثال در دایره‌های دو‌تایی به 4 بخش: دایره A، دایره B، اشتراک و اجتماع آن که جزئی از A و B است و در انتها بنا به تعریف مجموعه‌ی مرجع قرار دارد. A-B ( تعداد اعضای مجموعه) برابر است با : تعداد اعضای مجموعه اول - تعداد اعضای مشترک با B و همین قضیه برای اعضای B نیز صادق است، البته این اعداد فقط تعداد اعضایی که در A هست و در B نیست را توضیح می‌دهد.

<div id="probability_space" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=0f6d02 size=6>
فضای احتمال و پیشامد:
        </font>
		<p></p>
		<hr>
        فضای احتمال یا مدل احتمالاتی در یک آزمایش از عوامل زیر تشکیل می‌شود:
        <ul>
            <li>
           مجموعه $\Omega$ شامل کلیه نتایج ممکن برای آزمایش.
            </li>
            <li>
            زیر مجموعه‌های $\Omega$ که پیشامد نامیده می‌شوند.
            </li>
            <li>
           عدد $P(A)$ که به هر یک از پیشامد‌ها نسبت داده می شود.(طبق اصول موضوعه)
            </li>
        </ul>
             فضای نمونه $\Omega = (Sample Space)$ مجموعه کلیه نتایج ممکن برای یک آزمایش تصادفی است:   
        <ul>
            <li>
          برای مثال در آزمایش انداختن سکه داریم:      $\Omega = \{H , T\}$ که داری 2 عضو می‌باشد.
            </li>
            <li>
            برای مثالی دیگر در آزمایش پرتاب دو سکه داریم:        $\Omega = \{(H,H) , (H,T) , (T,H) , (T,T)\}$ که دارای 4 عضو می باشد.
            </li>
            <li>
            اگر آزمایش پرتاب 2 تاس باشد، فضای نمونه شامل 36 نقطه است.
            </li>   
            <li>
            اگر آزمایش طول عمر(به ساعت) یک ترانزیستور باشد آنگاه فضای نمونه شامل تمام اعداد حقیقی نامنفی است:
            $S = \{x : 0 \leqslant x < \infty\}$
 </li>
        </ul>
        پیشامد(Event): هر زیر‌مجموعه از فضای نمونه‌ها را واقعه یا پیشامد می‌گویند.
        <ul>
            <li>
            در آزمایش انداختن دو سکه داریم: $A = \{(H,H) , (H,T) , (T,H)\}$
            که $A$ پیشامد این است که حداقل 1 شیر بیاید.         
            </li>      
        </ul>
       <ul>
            <li>
            برای مثالی دیگر در آزمایش پرتاب دو سکه داریم:        $B = \{(H,H) , (T,T)\}$ پیشامد شبیه هم آمدن دو سکه است.
            </li>        
            <li>
            در مثال طول عمر ترانزیستور $E = \{x : 0 \leqslant x < 5\}$ پیشامد آن است که طول عمر ترانزیستور از 5 بیشتر نباشد.
            </li>
      </ul>


<div id="probability_space" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
<br>
        پیشامد‌های $A$ و $B$ را جدا از هم یا ناسازگار می‌گوییم هر گاه مجموعه‌های $A$ و $B$ جدا از هم باشند:    $A \cap B = \varnothing $     
	</font>
</div>

<div id="Axiomatic" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#0f6d02 size=6>
تعریف $Axiomatic$:
        </font>
		<p></p>
		<hr>
        اصول کولموگروف: به هر پیشامد A عدد $P(A)$ نسبت داده می‌شود به طوری که:
        <ul>
            <li>
            اصل(1) : $P(A) \geqslant 0$ (احتمال هر پیشامد نامنفی است.)
            </li>
            <li>
            اصل(2) : $P(\Omega) = 1$ (احتمال رخداد کل فضای نمونه ۱ است.)
            </li>
            <li>
            اصل(3) : اگر واقعه‌های $A$ و $B$ ناسازگار باشند ($A \cap B = \varnothing$)، آنگاه: $P(A \cup B) = P(A) + P(B)$
        </ul>
        <br/>
        با تکرار اصل 3 می‌توان آن را برای هر تعداد محدودی از وقایع بیان کرد ولی برای تعداد نامحدود به جای اصل 3 اصل قوی‌تری را باید جایگزین کرد:
        <br/>
        اصل $(3')$: اگر $A_1 , A_2 , ... $ دو به دو ناسازگار باشند یعنی هیچ دوتایی امکان ندارد باهم رخ دهند، داریم:
        <br/>
        $P(A_1 \cup A_2 \cup ... ) = P(A_1) + P(A_2) + ... $که این را اصل جمع‌پذیری نامحدود(قابل شمارش) می‌گویند.
	</font>
</div>

<div id="subjective" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#0f6d02 size=6>
  احتمال به عنوان معیاری از میزان اعتقاد به یک امر(subjective)
</font>
<p></p>
		<hr>
        گاهی اوقات به آسمان نگاه می‌کنیم و می‌گوییم به احتمال $۹۰%$ درصد باران می‌بارد. احتمال‌هایی که در چنین جمله‌ای بدان اشاره می‌شود اندازه‌ی اعتقاد فرد گوینده را بیان می‌کند. روشن است که این تعریف وابسته به فرد است و نمی‌تواند مبنای یک تئوری ریاضی و مستحکم قرار گیرد. اما می‌توان نتیجه‌گیری‌های معقولانه هم داشت. به نظر منطقی می‌رسد که احتمال به اندازه اعتقاد نیز باید اصول احتمال را رعایت کند. مثلا اگر کسی می‌گوید $۷۰%$ درصد مطمئن هستیم که کتاب "خورشید به دور زمین می‌گردد" را امین نوشته است و $۲۰%$ درصد مطمئن هستیم که محمد نوشته است، منطقی است بگوییم به احتمال $۹۰%$ درصد یکی از این دو نفر کتاب را نوشته است.
        <br/>
        تعریف فراوانی(فرکانس) نسبی (تعریف آماری یا تجربی):
        <br/>
        اگر یک آزمایش تصادفی را n بار انجام دهیم و در این n بار، $n_A$ بار واقعه A اتفاق افتد، علی‌الاصول برای n بزرگ، $P(A)$ عددی نزدیک به $\frac{n_A}{n}$ خواهد بود، این دیگر subjective نیست و قابل بررسی تجربی است.
	</font>
</div>

<div id="probability_space" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>    
    <font color=95026b size=6>
        مثال:
</font>
<br>
       در یک مدرسه با ۴۰۰ دانش‌آموز؛ ۳۰۰ نفر از آن‌ها به رشته‌ی ریاضی یا تجربی و یاهردو علاقه مندند؛ ۱۶۰ نفر به تجربی و ۱۲۰ نفر هم به ریاضی و هم به تجربی علاقه دارند. احتمال اینکه دانش‌آموزی از مدرسه به تصادف انتخاب شود که به رشته‌ی ریاضی علاقه‌مند باشد چقدر است؟
     <br>
     <font color=95026b size=5>
        حل:
</font>  
      <br>
    A را پیشامد علاقه به تجربی و B را پیشامد علاقه به ریاضی بگیریم. 
        پس $P(A \cup B)=\frac{۳۰۰}{۴۰۰}$ و $P(A)=\frac{۱۶۰}{۴۰۰}$و$P(A \cap B)=\frac{۱۲۰}{۴۰۰}$ بنابراین از رابطه بالا نتیجه می‌شود:
        $$P(B) = P(A\cup B)+P(A\cap B)-P(A) = .۷۵+.۳-.۴ = .۶۵$$
<br>
<font color=95026b size=6>
       مثال:
</font>
<br>
احتمال این‌که یک هواپیمای جنگی پس از انجام ماموریت سالم به پایگاه برگردد $\frac{۳۴}{۳۵}$ است یعنی از هر ۳۵ ماموریت یک ماموریت با حادثه همراه است. یکی از خلبانان پایگاه معتفد است اگر خلبانی از ۳۴ ماموریت مداوم جان سالم به‌در برد و حادثه‌ای نداشت باید از ماموریت بعدی صرف نظر کند. آيا این شخص درست می‌گوید؟

وقایع مستقل و وابسته
        </font>
        <hr/>
      در مطالعه احتمالات معمولا با دو دسته از وقایع مواجه می‌شویم: وقایع مستقل و وقایع وابسته 
        <ul>
            <li>
           وقایع مستقل: به دو پیشامد مستقل می‌گوییم اگر وقوع یکی بر احتمال وقوع دیگری تاثیر نگذارد. برای مثال در پرتاب یک تاس و سکه، عددی که از تاس به دست می آید ارتباطی با این که سکه خط آمده یا شیر ندارد. به عبارت دیگر دو پیشامد $A$ و $B$ را مستقل می‌گوییم اگر: $P(A \cap B) = P(A).P(B)$
            </li>
            <li>
            وقایع وابسته: دو پیشامد را وابسته می‌گوییم اگر احتمال وقوع یکی بر احتمال وقوع دیگری تاثیر گذارد. در زندگی روزمره هم از این مفهوم استفاده می‌کنیم، مثلا دزدی کردن و به زندان رفتن، تلاش کردن و به موفقیت رسیدن و سرعت غیرمجاز رفتن و تصادف کردن و....
            </li>
            <li>
            نکته‌ی مهمی که باید به آن دقت شود این است که جدا یا ناسازگار بودن دو پیشامد لزوما به معنای مستقل بودن آن دو پیشامد نیست. برای روشن شدن این موضوع به مثال ساده زیر توجه کنید:
            <br/>
            در یک کیسه ۳۰ توپ داریم. ۱۷ تا از آن ها آبی، ۱۱ تا زرد رنگ و ۲ تای باقی‌مانده سبز هستند. حال یک توپ به تصادف از کیسه خارج می کنیم.
            <br/>
            i) احتمال این که توپ به رنگ آبی باشد برابر $\frac{۱۷}{۳۰}$ است.
            <br/>
            ii) احتمال این که این توپ به رنگ زرد باشد برابر $\frac{۱۱}{۳۰}$ است.
            <br/>
            iii) آیا دو پیشامد بالا ناسازگار هستند؟
            <br/>
            از آن جا که یک توپ نمی‌تواند هم آبی و هم زرد باشد، به عبارت دیگر اشتراک این دو پیشامد تهی است پس این دو پیشامد ناسازگار هستند.
            <br/>
            iv) آیا این دو پیشامد مستقل هستند؟ 
            <br/>
            جواب این سوال با توجه به به تعریف استقلال دو پیشامد منفی می‌باشد.
<br/>
            سوال:
            <br/>
            نشان دهید اگر دو پیشامد A و B مستقل باشند آنگاه $\overline{A}$ و $B$ نیز مستقل‌اند. 
            <br/>
            نشان دهید اگر دو پیشامد $A$ و $B$ مستقل باشند آنگاه $\overline{A}$ و $\overline{B}$ نیز مستقل اند.
            </li>
            <li>
            سه واقعه‌ی A و B و C را مستقل می‌گوییم هر گاه: 
           <br/>
            $P(A \cap B) = P(A).P(B)$   و
            <br/>
            $P(A \cap C) = P(A).P(C)$   و
            <br/>
            $P(B \cap C) = P(B).P(C)$   و
            <br/>
            $P(A \cap B \cap C) = P(A).P(B).P(C)$
            <br/>
            یعنی اگر سه واقعه مستقل باشند، دو به دو نیز مستقل‌اند ولی عکس آن لزوما صحیح نیست.
            <br/>
            <br/>
            سوال: نشان دهید اگر A و B و C مستقل باشند آنگاه A از $\overline{BC}$ و $B + C$ مستقل است.
            <br/>
            </li>
            <li>
            n واقعه $A_1$، $A_2$، ... و $A_n$ را مستقل گویند هرگاه برای هر دسته اعداد صحیح $k_1$، $k_2$، ... و $k_r$ ($r \leqslant n$) داشته باشیم:
            <br/>
            $P(A_{k_1} \cap A_{k_2} \cap ... \cap A_{k_r}) = P(A_{k_1})P(A_{k_2})...P(A_{k_r})$
            <br/>
            در واقع تعداد $2^{n} - (n + 1)$ رابطه باید برقرار باشد.
            </li>
        </ul>
        <br>
                در ادامه‌ی دوره با استقلال و وابستگی بیشتر آشنا خواهیم شد.
</font>
</div>

        <br/>
        یک دقیقه به ساعت ۱۲ شب مهره های ۱ تا ۱۰ را در جعبه قرار می دهیم و مهره شماره ۱۰ را خارج می کنیم(خارج کردن و گذاشتن مهره ها زمانی نمی برد). $\frac{۱}{۲}$ دقیقه به ۱۲ مهره های ۱۱ تا ۲۰ در جعبه قرار می گیرند و مهره شماره ۲۰ از جعبه خارج می شود. $\frac{۱}{۴}$ دقیقه به ۱۲ مهره های ۲۱ تا ۳۰ در جعبه قرار داده می شوند و مهره شماره ۳۰ از جعبه خارج می شود و الی آخر.
        <br/>
        حال سوال مورد نظر این است که ساعت ۱۲ چند مهره در جعبه قرار دارد؟
        <br/>
        در مرحله بعد آزمایش را به نحوی دیگر انجام می دهیم. دوباره همان جعبه و مهره ها را در نظر بگیرید.(جعبه در ابتدا خالی می باشد). حال ۱ دقیقه به ساعت ۱۲ مهره های شماره ۱ تا ۱۰ را در جعبه قرار می دهیم و مهره شماره ۱ را خارج می کنیم. $\frac{۱}{۲}$ دقیقه به ۱۲ مهره های ۱۱ تا ۲۰ را در جعبه قرار می دهیم و مهره شماره ۲ را خارج می کنیم. $\frac{۱}{۴}$ دقیقه به ۱۲ مهره های ۲۱ تا ۳۰ را در جعبه می گذاریم و مهره شماره ۳ را خارج می کنیم و الی آخر. 
        <br/>

        <br/>
        با این روند فکر می کنید در ساعت ۱۲ چند مهره در جعبه خواهد بود؟

        <br/>
        <ul>
            <li>
            $(A \cup B)\cap C = (A \cap B)\cup(B \cap C)$
            </li>
            <li>
            $(A \cap B)\cup C = (A \cup B)\cap(B \cup C)$
            </li>
            <li>
            $\overline{(A \cup B)} = \overline A \cap \overline B$
            </li>
            <li>
            $\overline{(A \cap B)} = \overline A \cup \overline B$
            </li>
        </ul>
   <font face="XB Zar" size=5>
    $\overline A $ به معنی مکمل مجموعه A است.
       <br>
        </font>
        با استفاده از اصول کولموگروف و قوانین مجموعه‌ها موارد زیر را اثبات کنید :
        <ul>
            <li>
            $P(\overline{A}) = 1 - P(A)$
            </li>
            <li>
            $P(\varnothing) = 0$
            </li>
            <li>
            $P(A) \leqslant  1$
            </li>
            <li>
            اگر $B \subset A$ باشد آنگاه: $P(B) \leqslant P(A)$.
            </li>
            <li>
            اگر $B \subset A$ باشد، آنگاه : $P(A - B) = P(A) - P(B)$
            </li>
            <li>
            برای هر دو زیر مجموعه $A$ و $B$ (نه لزوما ناسازگار) داریم: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
            </li>
            <li>
            $P(A \cap B) \geqslant P(A) + P(B) - 1$
            </li>
