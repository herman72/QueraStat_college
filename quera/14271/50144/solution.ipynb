مقدمه
        </font>
		<p></p>
		<hr>
        پیش از شروع احتمال ابتدا با روشهای ترکیبیاتی آشنا می‌شویم. در ادامه خواهیم دید که تسلط و آشنایی با این روش‌ها در محاسبه مسائل مختلف احتمالاتی اهمیت زیادی دارد. در این درسنامه با این روش‌ها آشنا می‌شویم.
علی پاهای خیلی بلندی دارد. بلندتر از همه بچه‌های دانشکده. او همیشه به جای آسانسور از پله‌ها استفاده می کند چون می تواند هر ۳ پله را یکی کند. او توانسته تعداد حالاتی که می‌تواند ۳ پله را بالا برود بصورت زیر رسم کند.
        <br>
        <img src="figs/rule_of_product.png" style="float:center; width:50%"/>
        <br>
        به علی کمک کنید تا بتواند برای حالت عمومی که تعداد پله ها $n$ تا باشد(با فرض اینکه او آنقدر تمرین کند تا بتواند همه $n$ پله را یکی کند) حساب کند که به چند طریق می تواند پله ها را بالا برود.

اصول شمارش
        </font>
		<p></p>
		<hr>
        فرض کنید بتوان از شهر A از طریق n جاده به شهر B رفت و از شهر B از طریق m جاده به شهر C رفت. اگر قرار باشد با عبور از B؛ از A به C سفر کنیم؛ آن گاه به ازای هر راهی که برای سفر از A به B انتخاب می‌کنیم برای سفر از B به C ؛m انتخاب داریم. در نتیجه در کل برای مسافرت از A به C از طریق B؛ m*n انتخاب وجود دارد. این مثال ساده ای از اصل ضرب است که تعریف آن در ادامه آمده.
        <br/>
        <br/>
         اصل ضرب: فرض کنید نحوه‌ی انجام کاری را بتوان به k مرحله تجزیه کرد. مرحله‌ی اول را به $n_1$ طریق بتوان انجام داد، مرحله دوم را به $n_2$ طریق بتوان انجام داد،... و مرحله k ام به $n_k$ طریق قابل انجام باشد. در این صورت کل کار را به $n_1\times n_2 \times ... \times n_k$ طریق می توان انجام داد.
         برای مثال اگر بخواهیم تعداد اعداد دورقمی فرد را پیدا کنیم می‌توان گفت برای نوشتن یک عدد دورقمی ابتدا رقم دهگان و سپس رقم یکان آن را بنویسیم. ۹ انتخاب برای رقم دهگان و به ازای هر انتخاب رقم دهگان ۵ انتخاب(۱و۳و۵و۷و۹) برای رقم یکان داریم پس تعداد اعداد دورقمی فرد می شود: 
        $9\times 5 = 45$
        <br/>
        <br/>
        اصل جمع: اگر برای انجام کار A لازم باشد کار های $A_1 , A_2 , ... , A_n$ را انجام داد و هیچ دوتایی از آن ها اشتراک نداشته باشند و کار $A_1$ را به $n_1$ طریق، کار $A_2$ را به $n_2$ راه مختلف، ... و کار $A_n$ را به $n_n$ طریق مختلف بتوان انجام داد آنگاه کار A را به $n_1 + n_2 + ... + n_n$ طریق مختلف می توان انجام داد.
        برای مثال برای به دست آوردن تعداد اعداد دورقمی زوج با ارقام متمایز می‌توان گفت برای رقم دهگان ۹ انتخاب داریم. تعداد انتخاب های رقم یکان بستگی به این دارد که رقم دهگان انتخاب شده زوج باشد یا فرد. پس مساله را به دوحالت بدون اشتراک تقسیم کرد: 
        ۱- رقم دهگان زوج باشد. در این حالت ۴ انتخاب برای رقم‌های دهگان و ۴ انتخاب برای یکان و طبق اصل ضرب ۱۶ عدد داریم. 
        ۲- رقم دهگان فرد باشد که در این حالت تعداد انتخاب های دهگان و یکان به ترتیب ۵ و ۵ است که باز طبق اصل ضرب می‌شود ۲۵ 
          و در مجموع می‌شود: 
$16+25 = 41$
        <br/>
        <br/>
        اصل متمم: فرض کنید B زیر مجموعه‌ای از مجموعه‌ی A باشد، در این صورت: $|A - B| = |A| - |B|$
         مثلا اگر بخواهیم همان مثال تعداد اعداد دورقمی زوج با ارقام متمایز که با اصل جمع حل کردیم را با کمک اصل متمم بدست آوریم می‌توان گفت اگر تعداد اعداد فرد دورقمی با رقم متمایز را داشته باشیم(۴۰ تا) تعداد اعداد زوج دورقمی با رقم متمایز می‌شود: تعداد همه‌ی اعداد دورقمی با ارقام متمایز منهای تعداد اعداد فرد دورقمی با ارقام متمایز یعنی: 
         $81-40 = 41$
	</font>
</div>

		<font color=95026b size=6>
            مثال:
        </font>
        <br>
        پویان n نفر از دوستانش را به جشن تولد خود دعوت کرده است. اگر همه‌ی دعوت‌شدگان در جشن شرکت کنند و هر یک از آن‌ها دقیقا یک بار با هر نفر دست دهد، در مجموع چندبار دست دادن اتفاق افتاده است؟
        <br>
        <font color=95026b size=5>
حل:
        </font>
        <br>
        n+1 نفر در جشن حضور دارند(با احتساب خود پویان) و هرنفر با n نفر دیگر دست می‌دهد پس تعداد افرادی که باهم دست می‌دهند روی هم $n(n+1)$ است. اما این تعداد بیش از مقدار واقعی است زیرا هر دست دادن دوبار شمرده شده است مثلا اگر نفر۱ با نفر۲ دست داده باشد این دست دادن هم در دست دادن های نفر۱ و هم در دست دادن های نفر۲ شمرده شده است در نتیجه تعداد واقعی می‌شود: $\frac{n(n+1)}{2}$

$۲^{n-۱} \times ۱ = ۲^{n-۱}$

جایگشت‌ها
        </font>
		<p></p>
		<hr>
        هر ترتیب(خطی) قرار گرفتن چند شی در کنار هم را یک جایگشت(خطی) از آن اشیاء می‌نامند.
        <br/>
        جایگشت r شی از n شی: اگر مجموعه‌ای از n شی در اختیار داشته باشیم، هر آرایش خطی متشکل از r تا از این اشیاء را یک جایگشت r شی از این n شی می‌نامیم. 
        به جایگشت r از n تبدیل r از n نیز می‌گویند و آن را با نماد $P(n , r)$ یا $P_r^n$ می‌دهند.
        <br/>
        <br>
        <font color=95026b size=6>
            مثال
        </font>
        <br>
         با حروف $A, B, C, D$ چند کلمه ۴ حرفی می‌توان ساخت؟
        <br/>
                <font color=95026b size=5>
حل:
        </font>
        <br>
        برای حرف اول کلمه ۴ انتخاب داریم. در جایگاه دوم حرفی که در جایگاه اول گذاشتیم را نمی‌توانیم انتخاب کنیم پس ۳ انتخاب داریم. به همین ترتیب برای جایگاه سوم، ۲ انتخاب و برای جایگاه چهارم، ۱ انتخاب داریم. پس جواب سوال برابر است با: $۴ \times ۳ \times ۲ \times ۱ $.
        <br>
                <img src="figs/abcd.gif" style="float:center; width:20%"/>
        <br>
        <br/>
        <font color=95026b size=6>
            مثال
        </font>
        <br>
       ۸ توپ قرمز و ۴ توپ سبز را می‌خواهیم در یک ردیف کنار هم بچینیم. به چند طریق می‌توان این کار را انجام داد به طوری که هیچ دو توپ سبزی کنار هم نباشند؟
        <br/>
        <font color=95026b size=5>
حل:
        </font>
        <br>
       ابتدا به $۸!$ طریق توپ‌های قرمز را می‌چینیم. حال ۹ جایگاه مجاز(نخستین جایگاه مجاز سمت چپ، چپ‌ترین توپ قرمز تا نهمین جایگاه سمت راست، راست‌ترین توپ قرمز) داریم که ۴ توپ سبز در آن ها قرار می‌دهیم که برای قراردادن سبز اول ۹ انتخاب و برای دومین سبز ۸ انتخاب و.... و برای چهارمین سبز ۶ انتخاب داریم. پس جواب برابر است با: $۸! \times (۹ \times ۸ \times ۷ \times ۶)$
	</font>

<div id="per2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#0f6d02 size=6>
جایگشت با تکرار
        </font>
		<p></p>
		<hr>
        تاکنون در مسائل مطرح شده صحبت از جایگشت‌هایی بوده که اشیاء در آن متمایز بوده‌اند. اما همیشه اشیاء مورد بحث همگی متمایز نیستند.
        <br/>
به‌عنوان یک مثال خیلی ساده برای درک مبحث، فرض کنید مریم می‌خواهد اسباب بازی‌های خود را که در شکل زیر می‌بینید در قفسه‌اش بچیند. به چند طریق می‌تواند این کار را بکند؟
<br>
<img src="figs/ornaments.png" style="float:center; width:40%"/>
<br>
بیایید در ابتدا فرض کنیم این اشیا همه متمایز‌اند. پاسخ می‌شود $۸!$. حال هر کدام از عروسک‌های مشابه اگر متمایز بودند  چند جایگشت داشتند؟
 سگ $۳!$ و گربه $۲!$
این حالات در واقع چون حالات تکراری می‌باشند نباید در محاسبه‌ی اولیه‌ی ما محاسبه می‌شد(به عبارت دیگر نتیجه‌ی جابه‌جا کردن ترتیب قرارگیری اشیاء مشابه را نباید جایگشت جدید به حساب آوریم). بنابر اصل ضرب این حالات قبلا در حاصل کل ضرب شده‌اند، پس کافی است ما 
حاصل قبلی را تقسیم بر تعداد حالات مربوط به موارد تکراری کنیم:
        $\frac{۸!}{۲! ۳!}$
<br>
        <br/>
        بصورت کلی
        n شی در نظر بگیرید طوری که $k_1$ شی آن با یک دیگر، $k_2$ شی دیگر با هم، ... و $k_t$ شی دیگر با هم مشابه باشند طوری که $k_1 + k_2 + ... + k_t = n$ شود. در آن صورت تعداد جایگشت‌های متمایز n شی برابر است با:
        $$p(n;k_1,k_2,.....,k_t)=\frac{n!}{k_1! k_2! ... k_t!}$$
        <br>
        معمولا در کتاب ها ترتیب(جایشگت) را با علامت p نمایش می‌دهند. این p را با  P به معنی احتمال که به زودی معرفی خواهیم کرد اشتباه نکنید!
	</font>
</div>

<div id="h2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=95026b size=6>
مثال
        </font>
		<br>
در چند جایگشت از حروف کلمه massasauga هیچ دوحرفی از سه حرف m و u و g مجاور نیستند؟
<br>
        <font color=95026b size=5>
حل
        </font>
        <br>
        ابتدا تمام حروف به غیر از این سه حرف را در یک ردیف قرار می‌دهیم. این کار به $\frac{۷!}{۳!۴!}$ طریق انجام می‌پذیرد(جایشگت ۷ شی که ۳تا و ۴تا یکسان هستند) اکنون در ۳ مکان از ۸ مکان بین بین ۷ حرف گذاشته شده باید ۳ حرف m و u و g را قرار دهیم. این کار نیز به $P(۸,۳)=\frac{۸!}{۵!}$ طریق می‌توان انجام داد. پس تعداد جایگشت های موردنظر برابر است با:
        $$\frac{7!8!}{3!4!5!}$$
</font>
</div>

<div id="per3" style="direction:rtl;line-height:300%;">

جایگشت دوری
        </font>
		<p></p>
		<hr>
        منظور از جایگشت دوری از n شی آرایشی از این n شی دور یک دایره است. با این ویژگی که اگر یک آرایش از دوران آرایش دیگری به دست آید آنگاه این دو آرایش را یکسان می‌گیریم. برای مثال در شکل زیر هر سه دایره یک جایگشت دوری به حساب می‌آید زیرا در هر سه ترتیب A وB وC نسبت به هم یکی است.
        <br/>
        در واقع در جایگشت دوری موقعیت اشیاء نسبت به هم اهمیت دارد.
        <br/>
<br>
<img src="figs/circular_p.png" style="float:center; width:40%"/>
<br>
        قضیه: تعداد جایگشت‌های دوری n شی متمایز برابر $(n - 1)!$ است.(چرا؟)
        <br/>
        <br>
        		<font color=95026b size=6>
مثال
        </font>
		<br>
         ۵ معلم مدرسه می‌خواهند با ۵ عضو شورای دانش‌آموزی مدرسه جلسه‌ای بگذارند. جلسه دور میزی با ۱۰ صندلی برگزار می‌شود. معلمان می‌خواهند بدانند چند حالت نشستن روی صندلی‌ها وجود دارد که هیچ دو دانش‌آموزی کنارهم نشینند؟
        <br/>
        <font color=95026b size=5>
حل
        </font>
        <br>
        ابتدا قراردادن ۵ معلم دور میز یک جایگشت دوری است که به $۴!$ معلم‌ها را دور میز می‌نشانیم.(فرض می‌کنیم دور میز در ابتدا خالی است و هر کس وقتی می‌خواهد در جایی بنشیند یک صندلی برای خودش می‌آورد) سپس در ۵ جای خالی بین معلمان ۵ دانش آموز را به $۵!$ طریق می‌نشانیم. چون معلم‌ها در مرحله قبل ثابت شده بودند پس گویی انتخاب‌های دانش آموزان مشابه جایگشت خطی می‌شود.
        <br/>
        پس جواب برابر است با: $4! \times 5!$
	</font>

           در بسیاری از مسائل ترکیبیاتی؛ برخلاف جایشگت‌ها ترتیب قرارگیری عناصر اهمیتی ندارد
       <br>
       تعریف ترکیب: اگر A مجموعه باشد، یک ترکیب A یعنی زیر مجموعه‌ای از A. به بیان دیگر یک آرایش نامرتب r شی از مجموعه A که شامل n شی است یک ترکیب r از n می‌گوییم.
        <br/>
        تعداد زیر مجموعه‌های r عضوی از مجموعه‌ای n عضوی(ترکیب r از n) را با نماد $n \choose r$ نمایش می‌دهیم.
        <br/>
        اگر $ 0 \leq r \leq n$ باشد، آنگاه:
        ${n \choose r} = \frac{n!}{r!(n - r)!}$
        <br>
        <br/>
        		<font color=95026b size=6>
مثال
        </font>
		<br>
         n نقطه روی محیط دایره انتخاب می‌کنیم. همه وتر‌های بین هر جفت از این نقاط را رسم می‌کنیم. با فرض این که هیچ 3 وتری در یک نقطه همرس نیستند تعداد نقاط تقاطع وتر‌ها درون دایره را بیابید.
        <br/>
                <font color=95026b size=5>
حل
        </font>
        <br>
        هر 4 نقطه ار بین n نقطه یک نقطه تقاطع متمایز تولید می‌کنند و هر نقطه تقاطع از تقاطع دو وتر تشکیل می‌شود که انتهای دو وتر 4 نقطه متمایز از n نقطه است. پس تعداد نقاط تقاطع برابر است با: $n \choose 4$
</font>
